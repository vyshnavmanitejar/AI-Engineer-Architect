## 📖 Introduction

Comments are **essential for writing readable and maintainable code**. In Python, comments are **ignored by the interpreter** and serve to explain the code to developers.

### Why Comments Matter?
✅ Improve code readability  
✅ Document logic and intentions  
✅ Help team collaboration  
✅ Simplify debugging and maintenance  

---

## 🔤 Types of Comments in Python

| Type | Symbol | Usage |
|------|--------|-------|
| **Single-Line** | `#` | Brief explanations |
| **Multi-Line** | Multiple `#` | Detailed explanations |
| **Docstrings** | `""" """` or `''' '''` | Function/class documentation |

---

## 1️⃣ Single-Line Comments

Single-line comments start with the `#` symbol and extend to the end of the line. They are used for **brief explanations or notes**.

### Characteristics
- Begins with `#`
- Extends to end of line
- Perfect for quick notes
- Most commonly used

### 📝 **Example 1: Simple Calculation**
```python
# Calculate the sum of two numbers
x = 5
y = 10
sum_result = x + y  # Store the result in sum_result
```

### 📝 **Example 2: Loop with Comment**
```python
# Print numbers from 1 to 5
for i in range(1, 6):
    print(i)  # Display each number
```

### 📝 **Example 3: Conditional Statement**
```python
# Check if user is an adult
age = 18
if age >= 18:  # Adult condition
    print("You are an adult")
```

---

## 2️⃣ Multi-Line Comments

Multi-line comments use **multiple `#` symbols** on consecutive lines. They are used for **extended explanations and documentation**.

### Characteristics
- Each line starts with `#`
- Used for detailed explanations
- Useful for complex logic
- More structured appearance

### 📝 **Example 1: Function Purpose**
```python
# This function calculates the factorial of a number
# It takes a positive integer as input
# Returns the factorial value
def factorial(n):
    if n == 1:
        return 1
    else:
        return n * factorial(n - 1)
```

### 📝 **Example 2: Algorithm Explanation**
```python
# Algorithm to find the largest number in a list
# Step 1: Initialize max with first element
# Step 2: Iterate through remaining elements
# Step 3: Update max if current element is greater
numbers = [3, 7, 2, 9, 1]
max_num = numbers[0]
for num in numbers[1:]:
    if num > max_num:
        max_num = num
```

### 📝 **Example 3: Code Section Header**
```python
# ==============================================
# DATA VALIDATION SECTION
# ==============================================
# Validate user input and handle edge cases
# Check for empty strings, null values, etc.

def validate_email(email):
    return "@" in email and "." in email
```

## 3️⃣ Docstrings (Documentation Strings)

Docstrings use **triple quotes** (`""" """` or `''' '''`) and are used to document **functions, classes, and modules**.

### Characteristics
- Enclosed in triple quotes
- Can be multi-line
- Accessible via `__doc__` attribute
- PEP 257 standard format
- Serves as inline documentation

### 📝 **Example 1: Function Docstring**
```python
def greet(name):
    """
    This function greets a person by name.
    
    Args:
        name: A string representing the person's name
    
    Returns:
        A greeting message as a string
    """
    return f"Hello, {name}!"

# Access docstring
print(greet.__doc__)
```

### 📝 **Example 2: Class Docstring**
```python
class Calculator:
    """
    A simple calculator class for basic arithmetic operations.
    
    Methods:
        add(a, b): Returns the sum of two numbers
        subtract(a, b): Returns the difference of two numbers
        multiply(a, b): Returns the product of two numbers
    """
    
    def add(self, a, b):
        """Add two numbers and return the result."""
        return a + b
```

### 📝 **Example 3: Module Docstring**
```python
"""
Module: math_utilities.py

This module contains utility functions for mathematical operations.
It includes functions for basic arithmetic, statistics, and geometry.

Author: Developer
Date: 2026-03-30
Version: 1.0
"""

def square_root(n):
    """Calculate the square root of a number."""
    return n ** 0.5
```

## 🎯 Best Practices for Comments

| Practice | Description |
|----------|-------------|
| **Be Clear** | Write comments that are easy to understand |
| **Stay Current** | Update comments when code changes |
| **Avoid Obvious** | Don't comment on self-explanatory code |
| **Use Proper Grammar** | Write professional, clear comments |
| **Keep It Short** | Use concise explanations |

---

## 📊 Comparison Table

| Feature | Single-Line | Multi-Line | Docstring |
|---------|------------|-----------|-----------|
| **Symbol** | `#` | Multiple `#` | `""" """` |
| **Length** | One line | Multiple lines | Multiple lines |
| **Accessibility** | No | No | Yes (`__doc__`) |
| **Best For** | Quick notes | Detailed notes | Function/class docs |
| **Professional Use** | ⭐⭐⭐⭐ | ⭐⭐⭐ | ⭐⭐⭐⭐⭐ |

## ✨ Key Takeaways

- **Single-line comments**: Use `#` for brief explanations
- **Multi-line comments**: Use multiple `#` for detailed explanations  
- **Docstrings**: Use `""" """` for function/class documentation
- **Good comments** make code maintainable and professional
- **Comment wisely** – only when necessary for clarity
