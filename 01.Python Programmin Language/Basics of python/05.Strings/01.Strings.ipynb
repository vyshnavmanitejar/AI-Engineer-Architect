## 🎯 What are Strings?

A **string** is a **sequence of characters** enclosed in quotation marks. Strings are one of the most important data types in Python.

### Characteristics
- **Immutable** (cannot be changed after creation)
- **Ordered** (each character has a position)
- **Can contain** letters, numbers, symbols, spaces
- **Created with quotes** - single `' '` or double `" "`

### Why Strings Matter?
✅ Handle user input and text data  
✅ Display messages and feedback  
✅ Process and analyze text  
✅ Essential for most programs  

---

## 📝 Creating Strings

Strings in Python are **surrounded by quotation marks**. You can use either **single quotes** `' '` or **double quotes** `" "`.

### Single Quotes vs Double Quotes

Both are equivalent - choose one and **be consistent**:
```python
message1 = 'Hello'          # Single quotes
message2 = "Hello"          # Double quotes
# Both create identical strings!
```

### When to Use Which?

| Situation | Use | Example |
|-----------|-----|---------|
| **Contains apostrophe** | Double quotes | `"It's great"` |
| **Contains double quote** | Single quotes | `'She said "Hi"'` |
| **When in doubt** | Either | Be consistent |

### Best Practice
Choose a style and **use it consistently throughout your code**. Python convention prefers **double quotes**.

````python
# Creating Strings - Basic Examples

# Example 1: Single and Double Quotes
message1 = 'Hello'
message2 = "Hello"
print("Single quotes:", message1)
print("Double quotes:", message2)
print()

# Example 2: Strings with Quotes Inside
sentence1 = "It's a great day"
sentence2 = 'He said "Hello"'
print(sentence1)
print(sentence2)
print()

# Example 3: Assigning to Variables
name = "Alice"
greeting = 'Welcome!'
print(greeting, name)
````

## 📦 Multiline Strings

You can create **strings that span multiple lines** using **three quotes** (triple quotes).

### Using Triple Double Quotes
```python
text = """This is a
multiline string
spanning multiple lines"""
```

### Using Triple Single Quotes
```python
text = '''This is also a
multiline string
with triple quotes'''
```

### Important
- **Line breaks are preserved** exactly as written in the code
- Useful for **long text, documentation, and multi-line content**
- Triple quotes can also be used for **comments and docstrings**

### When to Use
✅ Long text content  
✅ Poems, stories, documentation  
✅ SQL queries or code blocks  
✅ Preserving formatting

````python
# Multiline Strings Examples

# Example 1: Triple Double Quotes
poem = """Roses are red,
Violets are blue,
Python is awesome,
And so are you!"""
print("Poem:")
print(poem)
print()

# Example 2: Triple Single Quotes
note = '''This is a
multiline note
with three lines'''
print("Note:")
print(note)
print()

# Example 3: Multiline preserves formatting
address = """Alice Johnson
123 Main Street
New York, NY 10001"""
print("Address:")
print(address)
````

## 🔍 String Indexing (Accessing Characters)

**Strings are sequences of characters** that can be accessed individually using **index numbers**.

### How Indexing Works

Each character in a string has a **position** (index):
- **First character** is at index **0** (not 1!)
- **Last character** is at index **-1**
- **Second-to-last** is at index **-2**, and so on

### Visual Example
```
String:  "P  y  t  h  o  n"
Index:    0  1  2  3  4  5    (positive)
Index:   -6 -5 -4 -3 -2 -1    (negative)
```

### Accessing Characters

| Code | Result | Description |
|------|--------|-------------|
| `text[0]` | First character | Positive indexing |
| `text[2]` | Third character | Positive indexing |
| `text[-1]` | Last character | Negative indexing |
| `text[-2]` | Second-to-last | Negative indexing |
| `text[1:4]` | Characters at positions 1, 2, 3 | Slicing |

### Important Notes
- Strings are **ordered sequences**
- Accessing by index is **fast and efficient**
- Negative indices count from the **end backwards**
- Index must exist, or you get an **IndexError**

````python
# String Indexing Examples

# Example 1: Positive Indexing
text = "Python"
print("String:", text)
print("text[0] =", text[0])
print("text[1] =", text[1])
print("text[2] =", text[2])
print()

# Example 2: Negative Indexing
print("Negative Indexing:")
print("text[-1] =", text[-1], "(last character)")
print("text[-2] =", text[-2], "(second-to-last)")
print("text[-6] =", text[-6], "(first character)")
print()

# Example 3: Slicing (getting substrings)
message = "Hello, World!"
print("String:", message)
print("message[0:5] =", message[0:5])
print("message[7:12] =", message[7:12])
print("message[0:11] =", message[0:11])
print()

# Example 4: Different strings
name = "Alice"
print(f"First letter of '{name}' = {name[0]}")
print(f"Last letter of '{name}' = {name[-1]}")
````

## 📏 String Length

The **`len()` function** returns the **number of characters** in a string, including spaces and punctuation.

### Syntax
```python
len(string)
```

### Returns
- An **integer** representing the total character count
- **Includes spaces**, punctuation, and special characters
- Empty string `""` has length **0**

### Uses
✅ Validate input length  
✅ Loop exactly through all characters  
✅ Check if string is empty  
✅ Extract substrings by calculating position  

### Examples

| Code | Result | Description |
|------|--------|-------------|
| `len("Hello")` | `5` | 5 characters |
| `len("Hi!")` | `3` | Includes punctuation |
| `len("Hi ")` | `3` | Includes spaces |
| `len("")` | `0` | Empty string |
| `len("Python Programming")` | `18` | Includes space |

````python
# String Length Examples

# Example 1: Basic string length
text1 = "Hello"
text2 = "Python"
text3 = "Hello, World!"
print("String Length:")
print(f'len("{text1}") = {len(text1)}')
print(f'len("{text2}") = {len(text2)}')
print(f'len("{text3}") = {len(text3)}')
print()

# Example 2: Length with spaces
text_with_space = "Hello World"
text_without_space = "HelloWorld"
print("Spaces are counted:")
print(f'len("Hello World") = {len(text_with_space)}')
print(f'len("HelloWorld") = {len(text_without_space)}')
print()

# Example 3: Empty string
empty = ""
print(f'len("") = {len(empty)} (empty string)')
print()

# Example 4: Using length in loops
word = "python"
print(f"Characters in '{word}':")
for i in range(len(word)):
    print(f"Position {i}: {word[i]}")
````

## 🔄 Looping Through Strings

Since **strings are sequences of characters**, you can **iterate** (loop) through each character using a **`for` loop**.

### Syntax
```python
for variable in string:
    # Do something with each character
```

### How It Works
- **Each iteration** processes one character
- Loop continues until all characters are processed
- Access the current character using the variable

### Common Uses
✅ Process each character one by one  
✅ Count specific characters  
✅ Validate or transform text  
✅ Search for patterns  

### Example
```python
text = "Hello"
for letter in text:
    print(letter)

# Output:
# H
# e
# l
# l
# o
```

### With Index
Use **`enumerate()`** to get both index and character:
```python
text = "Hello"
for index, letter in enumerate(text):
    print(f"Position {index}: {letter}")
```

````python
# Looping Through Strings Examples

# Example 1: Simple loop
word = "Python"
print("Example 1 - Characters in 'Python':")
for letter in word:
    print(letter, end=" ")
print("\n")

# Example 2: Loop with operations
text = "hello"
print("Example 2 - Uppercase each letter:")
for letter in text:
    print(letter.upper(), end="")
print("\n\n")

# Example 3: Using enumerate (with index)
word = "Code"
print("Example 3 - With indices:")
for index, letter in enumerate(word):
    print(f"Position {index}: {letter}")
print()

# Example 4: Count specific characters
text = "banana"
count = 0
print("Example 4 - Count 'a' in 'banana':")
for letter in text:
    if letter == "a":
        count += 1
print(f"Letter 'a' appears {count} times")
````

## 🔎 Checking if String Contains Text

The **`in` keyword** checks if a **substring exists** in a string.

### Syntax
```python
substring in string      # Returns True or False
```

### Returns
- **`True`** if substring is found
- **`False`** if substring is not found
- **Case-sensitive** (e.g., "Python" ≠ "python")

### Common Uses
✅ Validate user input  
✅ Search for keywords  
✅ Filter strings  
✅ Make decisions based on content  

### Examples

| Code | Result | Description |
|------|--------|-------------|
| `"ell" in "hello"` | `True` | Substring exists |
| `"xyz" in "hello"` | `False` | Not found |
| `"Hello" in "hello"` | `False` | Case matters |
| `"hello" in "Hello World"` | `False` | Case matters |

### Using in Conditions
```python
text = "Python is awesome"
if "awesome" in text:
    print("Found!")
```

````python
# Checking String Content - Using 'in'

# Example 1: Simple check
text = "Python is awesome"
print("Example 1 - Basic checks:")
print(f"'is' in text = {'is' in text}")
print(f"'xyz' in text = {'xyz' in text}")
print()

# Example 2: Case sensitivity
phrase = "Hello, World!"
print("Example 2 - Case sensitivity matters:")
print(f"'world' in phrase = {'world' in phrase}")
print(f"'World' in phrase = {'World' in phrase}")
print()

# Example 3: Using in if statements
sentence = "The best things in life are free!"
print("Example 3 - Using in conditions:")
if "free" in sentence:
    print("✅ 'free' is present in the sentence")
else:
    print("❌ 'free' is not present")
print()

# Example 4: Searching in different strings
email = "john@example.com"
print("Example 4 - Search for '@' in email:")
if "@" in email:
    print(f"'{email}' is a valid email format")
else:
    print("Invalid email format")
````

## ❌ Checking if String Does NOT Contain Text

The **`not in` keyword** checks if a **substring does NOT exist** in a string.

### Syntax
```python
substring not in string      # Returns True or False
```

### Returns
- **`True`** if substring is NOT found
- **`False`** if substring is found
- **Opposite of `in`**
- **Case-sensitive**

### Common Uses
✅ Validate forbidden content  
✅ Filter unwanted words  
✅ Check for errors or warnings  
✅ Exclusion logic  

### Examples

| Code | Result | Description |
|------|--------|-------------|
| `"xyz" not in "hello"` | `True` | Not found |
| `"ell" not in "hello"` | `False` | Found (so NOT True) |
| `"Python" not in "java"` | `True` | Not found |

### Using not in Conditions
```python
text = "I like apples"
if "oranges" not in text:
    print("Apples are preferred!")
```

### Combining with if/else
```python
keyword = "error"
message = "Something went wrong"

if keyword not in message:
    print("No errors found")
else:
    print("Error detected!")
```

````python
# Checking if NOT in String - Using 'not in'

# Example 1: Basic 'not in' checks
text = "Python is awesome"
print("Example 1 - Basic 'not in' checks:")
print(f"'xyz' not in text = {'xyz' not in text}")
print(f"'awesome' not in text = {'awesome' not in text}")
print()

# Example 2: Using in if statements
sentence = "The best things in life are free!"
print("Example 2 - Using 'not in' in conditions:")
if "expensive" not in sentence:
    print("✅ 'expensive' is NOT in the sentence")
else:
    print("❌ 'expensive' IS in the sentence")
print()

# Example 3: Validating forbidden content
password = "MyPassword123"
print("Example 3 - Checking for forbidden characters:")
if " " not in password:
    print("✅ No spaces - Password is valid")
else:
    print("❌ Spaces not allowed in password")
print()

# Example 4: Combining 'in' and 'not in'
email = "john@example.com"
print("Example 4 - Validating email:")
if "@" in email and "." in email:
    if " " not in email:
        print(f"✅ '{email}' looks like a valid email")
    else:
        print("❌ Email contains spaces")
else:
    print("❌ Invalid email format")
````

## 🎯 Key Takeaways - Strings Recap

### String Creation

| Method | Example | Use Case |
|--------|---------|----------|
| **Single quotes** | `'Hello'` | Simple strings |
| **Double quotes** | `"Hello"` | Contains apostrophes |
| **Triple quotes** | `"""Multi\nline"""` | Long text, preserving format |

### String Operations Summary

| Operation | Syntax | Result | Example |
|-----------|--------|--------|---------|
| **Length** | `len(s)` | Integer | `len("Hello")` → `5` |
| **Access** | `s[index]` | Character | `"Hello"[0]` → `"H"` |
| **Slice** | `s[start:end]` | Substring | `"Hello"[1:4]` → `"ell"` |
| **Loop** | `for c in s:` | Iterate | Through each character |
| **Contains** | `sub in s` | Boolean | `"ell" in "Hello"` → `True` |
| **Not contains** | `sub not in s` | Boolean | `"xyz" in "Hello"` → `False` |

### Indexing Reference

```
String:  H  e  l  l  o
Positive: 0  1  2  3  4
Negative: -5 -4 -3 -2 -1
```

### Important Concepts

📌 **Zero-based indexing** - First character is at index `0`

📌 **Negative indexing** - `s[-1]` gets last character

📌 **Case-sensitive** - `"Hello" ≠ "hello"`

📌 **Immutable** - Once created, strings cannot be changed

📌 **Strings are sequences** - Can loop and index them

### Common Patterns

```python
# Check if substring exists
if "keyword" in my_string:
    # Do something

# Validate content
if "@" in email and "." in email:
    # Valid email format

# Loop through characters
for char in text:
    print(char)

# Get substring
substring = text[2:5]

# Find length
length = len(text)
```
