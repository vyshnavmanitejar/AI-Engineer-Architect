# 🎨 Python String Formatting

## Introduction

String formatting is a technique to combine **strings with variables and values** in a readable and dynamic way. Without formatting, you cannot simply concatenate strings and numbers.

### The Problem with String Concatenation

```
age = 36
txt = "My name is John, I am " + age  # ❌ This causes a TypeError!
```

Each data type must be converted before combining, which is tedious and error-prone.

### Solutions for String Formatting

| Method | Introduced | Status | Syntax |
|--------|-----------|--------|--------|
| **F-Strings** | Python 3.6+ | ✅ Preferred | `f"text {variable}"` |
| `format()` | Python 3.0+ | Older | `"text {}".format(variable)` |
| `%` operator | Python 2 | Legacy | `"text %s" % variable` |

### Why Use String Formatting?

✅ **Type-safe**: Automatically converts values to strings  
✅ **Readable**: Clear structure for dynamic content  
✅ **Flexible**: Supports expressions and formatting  
✅ **Pythonic**: Modern and recommended approach

---

## 🔤 F-Strings Basics

**F-Strings** (formatted string literals) were introduced in **Python 3.6** and are the **preferred method** for string formatting.

### Syntax

To create an f-string:
1. Add the prefix `f` or `F` before the opening quote
2. Use **curly brackets `{}`** as placeholders for variables, expressions, or operations

### Basic Syntax
```
f"Your message here {variable}"
```

### Characteristics

- ✅ **Faster than other methods** (evaluated at runtime)
- ✅ **More readable** (variables visible in string)
- ✅ **Supports expressions** (math, function calls, etc.)
- ✅ **No type conversion needed** (automatic)
- ✅ **Cleanest and most modern** approach

````python
# Example 1: Basic f-string with variable
name = "John"
age = 36
print(f"My name is {name}, I am {age}")

# Example 2: Multiple variables
first_name = "Alice"
last_name = "Smith"
print(f"Full name: {first_name} {last_name}")

# Example 3: String concatenation vs f-string
score = 95
# Old way (error-prone):
# txt = "Score: " + score  # ❌ Error!
# F-string way:
print(f"Score: {score}")

# Example 4: Mix of text and variables
product = "Laptop"
price = 1200
print(f"The {product} costs ${price}")
````

## 📍 Placeholders in F-Strings

**Placeholders** are the `{}` sections inside an f-string where you insert variables, values, or expressions.

### Basic Placeholder Usage

A placeholder can contain:
- **Variables** - direct variable values
- **Expressions** - calculations and operations
- **Function calls** - methods and functions
- **Modifiers** - formatting instructions (optional)

### Placeholder Syntax

```
f"Text {content_here}"
```

Where `content_here` can be:
- A variable name: `{age}`
- An expression: `{x + y}`
- A method call: `{text.upper()}`
- With modifier: `{price:.2f}`

### Example Use Cases

| Use Case | Code | Result |
|----------|------|--------|
| Single variable | `f"{age}"` | `36` |
| Multiple variables | `f"{name}, {age}"` | `John, 36` |
| Expression | `f"{10 + 5}"` | `15` |
| List/Dict access | `f"{items[0]}"` | First item |

````python
# Example 1: Simple placeholder with variable
price = 59
txt = f"The price is {price} dollars"
print(txt)

# Example 2: Multiple placeholders
item = "Apple"
quantity = 5
cost = 15
print(f"I bought {quantity} {item}s for ${cost}")

# Example 3: Placeholder with string method
text = "python"
print(f"Uppercase: {text.upper()}")

# Example 4: Accessing list/dict elements
colors = ["red", "green", "blue"]
scores = {"math": 95, "english": 88}
print(f"\nFirst color: {colors[0]}")
print(f"Math score: {scores['math']}")
````

## 🎛️ Format Modifiers

**Modifiers** are optional formatting instructions added to a placeholder using a **colon `:` followed by a format type**.

### Syntax

```
f"{value:format_type}"
```

### Common Format Modifiers

| Modifier | Type | Purpose | Example |
|----------|------|---------|---------|
| `.2f` | Float | 2 decimal places | `f"{3.14159:.2f}"` → `3.14` |
| `d` | Integer | Integer value | `f"{3.7:d}"` → `3` |
| `.2%` | Percent | Percentage with 2 decimals | `f"{0.75:.2%}"` → `75.00%` |
| `>10` | Right-align | Right-aligned (10 chars) | `f"{value:>10}"` |
| `<10` | Left-align | Left-aligned (10 chars) | `f"{value:<10}"` |
| `^10` | Center-align | Centered (10 chars) | `f"{value:^10}"` |
| `,` | Thousands | Comma separator | `f"{1000000:,}"` → `1,000,000` |

### Most Common Use: Fixed Decimals

The **`.2f` modifier** is particularly useful for **currency and precise numbers**, ensuring exactly 2 decimal places.

```
f"{price:.2f}"  # Always shows 2 decimals
```

````python
# Example 1: Fixed decimal (most common for currency)
price = 59.5
txt = f"The price is {price:.2f} dollars"
print(txt)

# Example 2: Percentage formatting
completion = 0.85
print(f"Progress: {completion:.2%}")

# Example 3: Thousands separator (comma formatting)
population = 1234567
print(f"Population: {population:,}")

# Example 4: Alignment (spacing)
name = "Python"
print(f"Right-aligned:  '{name:>10}'")
print(f"Left-aligned:   '{name:<10}'")
print(f"Center-aligned: '{name:^10}'")

# Example 5: Different number formats
value = 42
print(f"Binary: {value:b}")  # Binary
print(f"Hexadecimal: {value:x}")  # Hex
print(f"Scientific: {value:e}")  # Scientific notation
````

## 🧮 Expressions in Placeholders

One of the most powerful features of f-strings is the ability to **execute Python expressions directly inside placeholders**.

### What Can Go Inside `{}`?

- ✅ **Simple calculations**: `{10 + 5}`, `{x * 2}`
- ✅ **String operations**: `{text.upper()}`, `{name.title()}`
- ✅ **Function calls**: `{len(text)}`, `{abs(-10)}`
- ✅ **List/Dict access**: `{items[0]}`, `{data['key']}`
- ✅ **Conditional expressions**: `{'Yes' if x > 0 else 'No'}`
- ✅ **List comprehensions**: `{[i*2 for i in range(3)]}`

### Syntax

```
f"Result: {expression}"
f"Double: {value * 2}"
f"Status: {'Active' if condition else 'Inactive'}"
```

### Benefits

✅ **Clean and readable** - logic visible in string  
✅ **No intermediate variables** needed  
✅ **Dynamic content** calculated on-the-fly  
✅ **Powerful for debugging** and logging

````python
# Example 1: Math operations in placeholder
quantity = 5
price = 20
print(f"Total cost: {quantity * price} dollars")

# Example 2: String methods in placeholder
text = "hello world"
print(f"Uppercase: {text.upper()}")
print(f"Title case: {text.title()}")

# Example 3: Built-in functions
numbers = [1, 2, 3, 4, 5]
print(f"Sum: {sum(numbers)}")
print(f"Average: {sum(numbers) / len(numbers):.2f}")

# Example 4: Conditional expression (ternary operator)
age = 15
status = "Adult" if age >= 18 else "Minor"
print(f"Age category: {status}")
# Can also be written directly:
print(f"Age {age} is: {'Adult' if age >= 18 else 'Minor'}")

# Example 5: Combining multiple operations
width = 10
height = 5
print(f"Rectangle area: {width * height} sq units")
print(f"Rectangle perimeter: {2 * (width + height)} units")
````

### F-String Essentials

| Concept | Syntax | Example |
|---------|--------|---------|
| Basic f-string | `f"text"` | `f"Hello {name}"` |
| Multiple variables | `f"{x} {y}"` | `f"{first} {last}"` |
| Expression | `f"{expr}"` | `f"{x + y}"` |
| Decimal format | `f"{x:.2f}"` | `f"{price:.2f}"` |
| Alignment | `f"{x:>10}"` | Right-align in 10 chars |

### Practical Examples

**Currency Formatting:**
```python
price = 19.5
print(f"Total: ${price:.2f}")  # Output: Total: $19.50
```

**Debugging Output:**
```python
x = 10
y = 20
print(f"{x=}, {y=}")  # Output: x=10, y=20
```

**Dynamic Content:**
```python
items = ["apple", "banana"]
print(f"You have {len(items)} items")
```

### Important Notes on F-Strings

🔹 **Requires Python 3.6+** - check your version  
🔹 **Fastest method** compared to `.format()` or `%`  
🔹 **Most readable** - variables visible in string  
🔹 **Most flexible** - supports any Python expression  
🔹 **Preferred by Python community** - use this method  

### Advanced Tips

- Use `:,` for thousands separator: `f"{1000000:,}"` → `1,000,000`
- Use `:.2%` for percentages: `f"{0.85:.2%}"` → `85.00%`
- Use alignment for formatting: `f"{text:^20}"` - centered in 20 chars
- Combine multiple modifiers: `f"{value:>10.2f}"` - right-aligned, 2 decimals
