# ✔️ Python Booleans

## Introduction

A **Boolean** is a data type that can only have one of two values: `True` or `False`. Booleans are fundamental to programming logic and decision-making.

### Key Concepts

**Boolean Value**: A value that is either `True` or `False`  
**Boolean Expression**: An expression that evaluates to a Boolean value  
**Conditional Logic**: Using Booleans to make decisions with `if` statements  

### Why Booleans Matter

✅ Control program flow with **conditional statements**  
✅ Make **yes/no decisions** in code  
✅ **Compare values** to create logic  
✅ Build **complex conditions** for validation  
✅ Return results from **validation functions**  

### Simple: True or False

```python
# Boolean values are simple
print(True)   # Output: True
print(False)  # Output: False
```

### Boolean Expressions

A **Boolean expression** is any comparison or evaluation that results in `True` or `False`:

```
10 > 9  # True (10 is greater than 9)
10 == 9  # False (10 is not equal to 9)
10 < 9  # False (10 is not less than 9)
```

````python
# Example 1: Basic boolean values
print("Direct boolean values:")
print(True)
print(False)

# Example 2: Comparison expressions (evaluate to True/False)
print("\nComparison expressions:")
print(10 > 9)   # True
print(10 == 9)  # False
print(10 < 9)   # False

# Example 3: Store boolean result in variable
is_older = 25 > 20
print(f"\nIs 25 > 20? {is_older}")

# Example 4: Use in if statement
age = 20
if age >= 18:
    print(f"Age {age} is an adult")
else:
    print(f"Age {age} is not an adult")
````

## 🔍 Comparison Operators

**Comparison operators** are used to compare values and create expressions that evaluate to Boolean results.

### Available Comparison Operators

| Operator | Name | Purpose | Example | Result |
|----------|------|---------|---------|--------|
| `==` | Equal to | Check if values are equal | `10 == 9` | `False` |
| `!=` | Not equal to | Check if values are different | `10 != 9` | `True` |
| `>` | Greater than | Check if left is greater | `10 > 9` | `True` |
| `<` | Less than | Check if left is smaller | `10 < 9` | `False` |
| `>=` | Greater than or equal to | Check if left is greater or equal | `10 >= 10` | `True` |
| `<=` | Less than or equal to | Check if left is smaller or equal | `10 <= 9` | `False` |

### Common Mistakes

❌ **Confusion**: Using `=` (assignment) instead of `==` (comparison)  
✅ **Correct**: `if x == 5:` (comparison)  
❌ **Wrong**: `if x = 5:` (assignment)

### Use Cases

✅ Validate user input  
✅ Compare values in decisions  
✅ Check ranges (age, scores, etc.)  
✅ Verify equality or inequality  
✅ Build conditional logic

````python
# Example 1: Equality operators (== and !=)
x = 10
y = 9
print(f"x == y: {x == y}")  # False (10 is not equal to 9)
print(f"x != y: {x != y}")  # True (10 is not equal to 9)

# Example 2: Greater/Less than operators
print(f"\nx > y: {x > y}")    # True (10 > 9)
print(f"x < y: {x < y}")    # False (10 < 9)

# Example 3: Greater/Less than or equal
print(f"\nx >= 10: {x >= 10}")  # True (equal)
print(f"x <= 9: {x <= 9}")      # False

# Example 4: Comparing strings
name1 = "Alice"
name2 = "Bob"
print(f"\n'{name1}' == '{name2}': {name1 == name2}")
print(f"'{name1}' != '{name2}': {name1 != name2}")

# Example 5: Comparing with variables
age = 25
print(f"\nAge: {age}")
print(f"age >= 18: {age >= 18}")  # True (adult)
print(f"age < 18: {age < 18}")    # False (not minor)
````

## 📊 The `bool()` Function

The **`bool()` function** evaluates any value or expression and returns the **Boolean equivalent** (`True` or `False`).

### Syntax

```python
bool(value)
```

### How It Works

Every value in Python can be converted to a Boolean. The rules are:

1. **Most values are `True`** - any value with content
2. **Some values are `False`** - empty values and special cases

### Characteristics

✅ **Converts any type** to Boolean  
✅ **Follows consistency rules** (empty = False, content = True)  
✅ **Useful for validation** and debugging  
✅ **Helps understand truthiness** of values

````python
# Example 1: bool() with different types
print("Using bool() on different types:")
print(f"bool('Hello'): {bool('Hello')}")  # True (non-empty string)
print(f"bool(15): {bool(15)}")            # True (non-zero number)
print(f"bool(0): {bool(0)}")              # False (zero)
print(f"bool(''): {bool('')}")            # False (empty string)

# Example 2: bool() with lists
print(f"\nbool([1, 2, 3]): {bool([1, 2, 3])}")  # True (non-empty)
print(f"bool([]): {bool([])}")                  # False (empty)

# Example 3: bool() with variables
x = "Hello"
y = 15
z = 0

print(f"\nbool(x='{x}'): {bool(x)}")
print(f"bool(y={y}): {bool(y)}")
print(f"bool(z={z}): {bool(z)}")

# Example 4: Using bool() in conditions
value = ""
if bool(value):
    print("Value is truthy")
else:
    print("Value is falsy")
````

## ✅ Truthy Values

**Most values evaluate to `True`** when converted to Boolean. These are called **truthy values**.

### Truthy Values Reference

| Type | Truthy Examples | Rule |
|------|-----------------|------|
| **Strings** | `"Hello"`, `"a"`, `" "` | Any non-empty string |
| **Numbers** | `1`, `42`, `-5`, `3.14` | Any non-zero number |
| **Lists** | `[1, 2, 3]`, `["a"]` | Any non-empty list |
| **Tuples** | `(1, 2)`, `("x",)` | Any non-empty tuple |
| **Dicts** | `{"a": 1}`, `{"key": "val"}` | Any non-empty dictionary |
| **Sets** | `{1, 2, 3}`, `{"a"}` | Any non-empty set |
| **Boolean** | `True` | The literal value True |

### The General Rule

**"If it has content, it's truthy"**

✅ Non-empty containers  
✅ Non-zero numbers  
✅ Non-empty strings  
✅ The value `True`  

### Use Cases

✅ Check if a list has items  
✅ Validate non-empty strings  
✅ Test non-zero values  
✅ Condition on data availability

````python
# Example 1: Strings (non-empty are truthy)
print("Truthy strings:")
print(f"bool('Hello'): {bool('Hello')}")      # True
print(f"bool('a'): {bool('a')}")              # True
print(f"bool(' '): {bool(' ')}")              # True (space is content)

# Example 2: Numbers (non-zero are truthy)
print("\nTruthy numbers:")
print(f"bool(1): {bool(1)}")          # True
print(f"bool(42): {bool(42)}")        # True
print(f"bool(-5): {bool(-5)}")        # True
print(f"bool(3.14): {bool(3.14)}")    # True

# Example 3: Collections (non-empty are truthy)
print("\nTruthy collections:")
print(f"bool([1, 2, 3]): {bool([1, 2, 3])}")        # True
print(f"bool(['apple']): {bool(['apple'])}")        # True
print(f"bool({{'key': 'value'}}): {bool({'key': 'value'})}")  # True

# Example 4: Using truthy values in if statements
items = [1, 2, 3]
if items:  # Implicit truthy check
    print(f"\nList has {len(items)} items")

name = "Alice"
if name:  # Implicit truthy check
    print(f"Hello, {name}!")
````

## ❌ Falsy Values

There are **very few values that evaluate to `False`**. These are called **falsy values** and consist mainly of empty or zero values.

### Complete List of Falsy Values

| Type | Falsy Examples | Reason |
|------|---|---|
| **Boolean** | `False` | The literal value False |
| **None** | `None` | Python's null value |
| **Number** | `0`, `0.0`, `0j` | Zero in any numeric form |
| **Strings** | `""` | Empty string (no content) |
| **Lists** | `[]` | Empty list |
| **Tuples** | `()`, `((),)` | Empty tuple |
| **Dicts** | `{}` | Empty dictionary |
| **Sets** | `set()` | Empty set |

### The General Rule

**"If it's empty or zero or None or False, it's falsy"**

❌ Empty containers  
❌ Zero (0, 0.0, 0j)  
❌ Empty strings  
❌ The value `False`  
❌ The value `None`  

### Important Note

🔹 Only these specific values are falsy  
🔹 Everything else (including `"0"`, `" "`, negative numbers) is **truthy**  
🔹 This is a **complete list** - there are only these falsy values

````python
# Example 1: Falsy values - the complete list
print("All falsy values:")
print(f"bool(False): {bool(False)}")        # False
print(f"bool(None): {bool(None)}")          # False
print(f"bool(0): {bool(0)}")                # False
print(f"bool(0.0): {bool(0.0)}")            # False
print(f"bool(''): {bool('')}")              # False (empty string)
print(f"bool([]): {bool([])}")              # False (empty list)
print(f"bool(()): {bool(())}")              # False (empty tuple)
print(f"bool({{}}): {bool({})}")            # False (empty dict)
print(f"bool(set()): {bool(set())}")        # False (empty set)

# Example 2: Surprising truthy values
print("\nSurprising truthy values:")
print(f"bool('0'): {bool('0')}")            # True! (string with content)
print(f"bool(' '): {bool(' ')}")            # True! (space is content)
print(f"bool(-0): {bool(-0)}")              # False (negative zero is still zero)
print(f"bool(-5): {bool(-5)}")              # True (negative numbers are truthy)

# Example 3: Using falsy values in conditions
items = []
if not items:  # Check if empty (falsy)
    print("\nList is empty")

value = None
if value is None:  # Explicit check
    print("Value is None")

# Example 4: Falsy in real code
password = ""
if password:  # Won't execute (empty string is falsy)
    print("Password provided")
else:
    print("Password is empty - please enter a password")
````

## 🔗 Logical Operators

**Logical operators** combine multiple Boolean expressions to create more complex conditions.

### Available Logical Operators

| Operator | Purpose | Syntax | Returns True When |
|----------|---------|--------|-------------------|
| `and` | Logical AND | `a and b` | **Both** conditions are True |
| `or` | Logical OR | `a or b` | **At least one** condition is True |
| `not` | Logical NOT | `not a` | The condition is False (negation) |

### Truth Tables

**AND operator:**
- `True and True` → `True`
- `True and False` → `False`
- `False and True` → `False`
- `False and False` → `False`

**OR operator:**
- `True or True` → `True`
- `True or False` → `True`
- `False or True` → `True`
- `False or False` → `False`

**NOT operator:**
- `not True` → `False`
- `not False` → `True`

### Use Cases

✅ Combine multiple conditions  
✅ Verify multiple requirements  
✅ Create complex validation logic  
✅ Build sophisticated decision rules

````python
# Example 1: AND operator (both must be True)
x = 10
y = 9
print("AND operator:")
print(f"x > 8 and y < 10: {x > 8 and y < 10}")    # True (both true)
print(f"x > 8 and y > 10: {x > 8 and y > 10}")    # False (second false)

# Example 2: OR operator (at least one must be True)
print("\nOR operator:")
print(f"x > 8 or y > 10: {x > 8 or y > 10}")      # True (first is true)
print(f"x < 5 or y < 10: {x < 5 or y < 10}")      # True (second is true)
print(f"x < 5 or y > 10: {x < 5 or y > 10}")      # False (both false)

# Example 3: NOT operator (negation)
print("\nNOT operator:")
print(f"not (x > 8): {not (x > 8)}")              # False (negates True)
print(f"not (x < 5): {not (x < 5)}")              # True (negates False)

# Example 4: Practical example - age validation
age = 25
has_license = True
can_drive = age >= 18 and has_license
print(f"\nAge: {age}, License: {has_license}")
print(f"Can drive: {can_drive}")

# Example 5: Complex conditions with multiple operators
temperature = 22
is_weekend = True
print(f"\nTemperature: {temperature}, Weekend: {is_weekend}")
go_outside = (temperature > 15) and (temperature < 30) and (is_weekend or True)
print(f"Good time to go outside: {go_outside}")
````

## 🔎 Built-in Functions Returning Booleans

Python provides many **built-in functions that return Boolean values**. These are essential for validation and type checking.

### Common Built-in Boolean Functions

| Function | Purpose | Returns | Example |
|----------|---------|---------|---------|
| `isinstance(obj, type)` | Check if object is of a type | Boolean | `isinstance(5, int)` → `True` |
| `len(x) > 0` | Check if container has items | Boolean | `len([1, 2]) > 0` → `True` |
| `x in y` | Check if item is in container | Boolean | `"a" in "apple"` → `True` |
| `callable(obj)` | Check if object is callable | Boolean | `callable(print)` → `True` |
| `hasattr(obj, attr)` | Check if object has attribute | Boolean | `hasattr([1, 2], 'append')` → `True` |

### Type Checking with `isinstance()`

```python
isinstance(value, type)
```

Common types to check:
- `int` - integers
- `float` - floating point numbers
- `str` - strings
- `list` - lists
- `dict` - dictionaries
- `bool` - Booleans

### Use Cases

✅ Validate input types  
✅ Check data before processing  
✅ Ensure safe operations  
✅ Type-specific branching  
✅ Error prevention

````python
# Example 1: isinstance() - check data types
x = 200
y = "hello"
z = [1, 2, 3]

print("Type checking with isinstance():")
print(f"isinstance(x, int): {isinstance(x, int)}")           # True
print(f"isinstance(y, str): {isinstance(y, str)}")           # True
print(f"isinstance(z, list): {isinstance(z, list)}")         # True
print(f"isinstance(x, str): {isinstance(x, str)}")           # False

# Example 2: Check multiple types (use tuple)
value = 5
print(f"\nIs value int or float? {isinstance(value, (int, float))}")

# Example 3: in operator - check membership
text = "apple"
print(f"\n'a' in '{text}': {'a' in text}")         # True
print(f"'z' in '{text}': {'z' in text}")           # False

items = [1, 2, 3, 4, 5]
print(f"3 in {items}: {3 in items}")               # True
print(f"10 in {items}: {10 in items}")             # False

# Example 4: callable() - check if object is callable
print("\nIs it callable?")
print(f"callable(print): {callable(print)}")       # True
print(f"callable(5): {callable(5)}")               # False
print(f"callable(len): {callable(len)}")           # True

# Example 5: len() for checking if container is non-empty
items = [1, 2, 3]
if len(items) > 0:
    print(f"\nList has {len(items)} items")

# Alternative: use truthiness directly
if items:  # Simpler!
    print(f"List is not empty (has {len(items)} items)")
````

## 🔄 Functions Returning Booleans

You can **create your own functions that return Boolean values**. This is useful for validation, checking conditions, and making code more readable.

### Why Return Booleans?

✅ Create reusable validation logic  
✅ Make code more readable (function name describes what it checks)  
✅ Separate concerns (validation logic in function)  
✅ Test conditions easily  
✅ DRY principle (Don't Repeat Yourself)

### Simple Function Pattern

```python
def is_valid_age(age):
    return age >= 18
```

### Using Boolean-Returning Functions

```python
def is_valid_age(age):
    return age >= 18

if is_valid_age(25):
    print("You are an adult")
```

### Real-World Examples

```python
def is_even(num):
    return num % 2 == 0

def is_valid_email(email):
    return "@" in email and "." in email

def is_strong_password(pwd):
    return len(pwd) >= 8 and any(c.isdigit() for c in pwd)
```

### Benefits

🔹 **Readable code** - `if is_valid_email(email):` is clearer than complex logic  
🔹 **Testable** - easy to unit test validation functions  
🔹 **Reusable** - use the function multiple times  
🔹 **Maintainable** - change logic in one place

````python
# Example 1: Simple validation function
def is_valid_age(age):
    return age >= 18

print("Simple validation function:")
print(f"is_valid_age(25): {is_valid_age(25)}")  # True
print(f"is_valid_age(15): {is_valid_age(15)}")  # False

# Example 2: Use in if statement
age = 30
if is_valid_age(age):
    print(f"\nAge {age} - You can vote")
else:
    print(f"\nAge {age} - Too young to vote")

# Example 3: More complex validation function
def is_even(num):
    return num % 2 == 0

numbers = [10, 15, 20, 25, 30]
print(f"\nEven numbers in {numbers}:")
for num in numbers:
    if is_even(num):
        print(f"  {num} is even")

# Example 4: Validation function with multiple conditions
def is_strong_password(password):
    has_length = len(password) >= 8
    has_digit = any(char.isdigit() for char in password)
    has_upper = any(char.isupper() for char in password)
    return has_length and has_digit and has_upper

passwords = ["abc", "Pass123", "password123", "PASS123"]
print("\nPassword validation:")
for pwd in passwords:
    strength = is_strong_password(pwd)
    status = "Strong" if strength else "Weak"
    print(f"  '{pwd}': {status}")

# Example 5: Function that uses isinstance()
def can_be_added(a, b):
    return isinstance(a, (int, float)) and isinstance(b, (int, float))

print("\nChecking if values can be added:")
print(f"can_be_added(5, 10): {can_be_added(5, 10)}")      # True
print(f"can_be_added(5, '10'): {can_be_added(5, '10')}")  # False
````

### Critical Rules

🔹 **Read left to right:** `a and b or c` reads as `(a and b) or c`
