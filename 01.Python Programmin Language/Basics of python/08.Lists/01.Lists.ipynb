# 🎯 Python Lists

## Introduction

**Lists** are fundamental Python data structures used to store **multiple items in a single variable**. They are one of 4 built-in collection types (Lists, Tuples, Sets, Dictionaries), each with distinct characteristics and use cases.

### What are Lists?

Lists are **ordered**, **changeable** (mutable), and **allow duplicate values**. They are created using square brackets `[]` and can store any data type—strings, integers, floats, booleans, or even mixed types.

### Key Characteristics

| Feature | Description |
|---------|-------------|
| **Ordered** | Items have a defined sequence; new items are added at the end |
| **Changeable** | You can modify, add, or remove items after creation |
| **Duplicates** | Lists can contain identical values |
| **Indexed** | Access items by position starting at index `[0]` |
| **Mixed Types** | Can store different data types in same list |

### Why Use Lists?

Lists are perfect for:
- Storing collections of related data (student names, scores, etc.)
- Maintaining order of items
- Modifying data dynamically
- Iterating through multiple elements
- Building data structures for real-world problems

````python
# Example 1: Basic list creation with strings
fruits = ["apple", "banana", "cherry"]
print("Fruits list:", fruits)

# Example 2: List with numbers
numbers = [10, 20, 30, 40, 50]
print("Numbers list:", numbers)

# Example 3: List with different data types
mixed_list = ["apple", 25, True, 3.14]
print("Mixed list:", mixed_list)

# Example 4: Empty list
empty_list = []
print("Empty list:", empty_list)

# Example 5: List with duplicate values
duplicates = ["apple", "banana", "apple", "cherry", "banana"]
print("List with duplicates:", duplicates)
````

## 🔑 List Items & Properties

### Indexing

List items are **indexed**, starting from `0` for the first item:

| Position | Item Value | Index |
|----------|-----------|-------|
| First | "apple" | `[0]` |
| Second | "banana" | `[1]` |
| Third | "cherry" | `[2]` |

### Ordered Behavior
Lists are **ordered**, meaning items maintain their sequence. When you add new items to a list, they're placed at the end. The order is preserved unless explicitly changed.

### Changeable (Mutable)
Lists are **changeable**, allowing you to:
- Modify existing items
- Add new items
- Remove items
- Rearrange items

This makes lists flexible for dynamic data management.

### Allow Duplicates
Lists can contain identical values at different indices. Duplicate items don't cause errors—they're treated as separate entries.

````python
# Example 1: Accessing items by index
fruits = ["apple", "banana", "cherry", "orange"]
print("First item (index 0):", fruits[0])
print("Third item (index 2):", fruits[2])

# Example 2: Negative indexing (from end)
print("Last item (index -1):", fruits[-1])
print("Second-to-last (index -2):", fruits[-2])

# Example 3: Modifying items
fruits[1] = "blueberry"
print("Modified list:", fruits)

# Example 4: Working with duplicates
colors = ["red", "blue", "red", "green", "red"]
print("Colors with duplicates:", colors)
print("Count of 'red':", colors.count("red"))

# Example 5: Adding items
fruits.append("mango")
print("After adding mango:", fruits)
````

## 📏 List Length

The **`len()`** function returns the number of items in a list. This is essential for:
- Validating list content
- Loop boundaries
- Checking if a list is empty
- Accessing the last valid index

### Using len()

```python
my_list = ["apple", "banana", "cherry"]
length = len(my_list)  # Returns 3
```

The returned value represents the **number of elements**, not the highest index (which would be `length - 1`).

````python
# Example 1: Basic length check
fruits = ["apple", "banana", "cherry"]
print("Number of items:", len(fruits))

# Example 2: Length changes as list grows
fruits.append("orange")
print("After adding one item:", len(fruits))

# Example 3: Empty list has length 0
empty_list = []
print("Empty list length:", len(empty_list))

# Example 4: Using length in loops
numbers = [10, 20, 30, 40, 50]
print("List has", len(numbers), "items")
for i in range(len(numbers)):
    print(f"  Position {i}: {numbers[i]}")

# Example 5: Length with mixed data types
mixed = ["text", 42, 3.14, True, None]
print("Mixed list length:", len(mixed))
````

## 🎨 List Data Types

### Lists Can Store Any Data Type

List items can be of **any Python data type**:

| Data Type | Example |
|-----------|---------|
| **String** | `["apple", "banana"]` |
| **Integer** | `[1, 5, 7, 9]` |
| **Float** | `[1.5, 2.7, 3.14]` |
| **Boolean** | `[True, False, False]` |
| **Mixed** | `["abc", 34, True, 3.14]` |

### Mixed-Type Lists

A single list can contain items of **different data types**. This flexibility makes Python lists powerful for storing heterogeneous collections:

```python
my_list = ["John", 25, 175.5, True]  # String, int, float, boolean
```

### Checking List Type

From Python's perspective, lists are **objects with data type `'list'`**. Use the `type()` function to verify:

```python
print(type(my_list))  # Output: <class 'list'>
```

````python
# Example 1: List of strings
fruits = ["apple", "banana", "cherry"]
print("String list:", fruits)

# Example 2: List of integers
numbers = [1, 5, 7, 9, 3]
print("Integer list:", numbers)

# Example 3: List of booleans
flags = [True, False, False, True]
print("Boolean list:", flags)

# Example 4: Mixed data types in one list
person = ["John", 28, 175.5, True]
print("Mixed list:", person)

# Example 5: Using type() to check list type
print("Type of fruits:", type(fruits))
print("Type of numbers:", type(numbers))

# Example 6: Checking individual item types
print("Type of first item (string):", type(person[0]))
print("Type of second item (int):", type(person[1]))
print("Type of fourth item (bool):", type(person[3]))
````

## 🔧 List Constructor

### Creating Lists with list()

In addition to using square brackets `[]`, you can create lists using the **`list()`** constructor function. This is useful when converting other data types to lists.

### Syntax

```python
# Using square brackets (most common)
my_list = ["apple", "banana", "cherry"]

# Using list() constructor
my_list = list(("apple", "banana", "cherry"))  # Note: double parentheses for tuple input
```

### When to Use list()?

- Converting other iterables (tuples, strings, ranges) to lists
- Creating an empty list: `empty = list()`
- Converting sequences into mutable lists
- More explicit/readable in some contexts

**Note:** Square brackets `[]` are the most Pythonic and commonly used method for list creation.

````python
# Example 1: Using list() constructor with tuple
original_tuple = ("apple", "banana", "cherry")
list_from_tuple = list(original_tuple)
print("List from tuple:", list_from_tuple)

# Example 2: Creating empty list with list()
empty_list = list()
print("Empty list:", empty_list)
print("Empty list length:", len(empty_list))

# Example 3: Using list() with range
numbers = list(range(1, 6))
print("List from range(1, 6):", numbers)

# Example 4: Converting string to list
word = "hello"
char_list = list(word)
print("List from string:", char_list)

# Example 5: Comparing methods
method1 = [1, 2, 3]           # Square brackets
method2 = list((1, 2, 3))     # list() constructor
print("Square brackets:", method1)
print("list() constructor:", method2)
print("Are they equal?", method1 == method2)
````

### Practice Tips

1. **Experiment with indexing** — Try positive and negative indices to understand position tracking
2. **Remember index starts at 0** — First item is at index `[0]`, not `[1]`
3. **Use len() before accessing** — Avoid index errors by checking list length
4. **Mix data types carefully** — Verify type before using type-specific operations (e.g., strings have different methods than integers)
5. **Practice list modification** — Get comfortable with appending, removing, and modifying items

### Common Mistakes to Avoid

⚠️ **Assuming 1-based indexing** — Python uses 0-based indexing  
⚠️ **Index out of range** — Accessing an index >= length causes errors  
⚠️ **Confusing `=` and `==`** — Use `=` for assignment, `==` for comparison  
⚠️ **Forgetting double parentheses in list()** — Use `list((1, 2, 3))` not `list(1, 2, 3)`
