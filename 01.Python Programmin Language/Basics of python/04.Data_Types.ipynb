## 🎯 What are Data Types?

**Data types** classify the type of value a variable stores. Different data types support different operations and have different behaviors.

### Why Data Types Matter?
- ✅ Determines what operations you can perform
- ✅ Affects how memory is allocated
- ✅ Influences how values are stored and accessed
- ✅ Essential for writing correct programs

### Python Data Types Overview

Python has **two main categories** of built-in data types:

#### **1. Scalar (Single Value) Data Types**
| Type | Example | Purpose |
|------|---------|---------|
| **String** | `"Hello"` | Text data |
| **Integer** | `42` | Whole numbers |
| **Float** | `3.14` | Decimal numbers |
| **Boolean** | `True` | True/False values |

#### **2. Collection (Multiple Values) Data Types**
| Type | Example | Purpose |
|------|---------|---------|
| **List** | `[1, 2, 3]` | Ordered, changeable, allows duplicates |
| **Tuple** | `(1, 2, 3)` | Ordered, unchangeable, allows duplicates |
| **Set** | `{1, 2, 3}` | Unordered, unique values, no duplicates |
| **Dictionary** | `{key: value}` | Ordered, key-value pairs |

## 🔍 Checking Data Type with `type()`

The **`type()` function** returns the data type of any variable.

### Syntax
```python
type(variable)
```

### Returns
- `<class 'str'>` for strings
- `<class 'int'>` for integers
- `<class 'float'>` for floats
- `<class 'bool'>` for booleans
- `<class 'list'>` for lists
- `<class 'tuple'>` for tuples
- `<class 'set'>` for sets
- `<class 'dict'>` for dictionaries

````python
# Type Checking with type() Function

# String example
name = "Alice"
print(f"name = {name}, type = {type(name)}")

# Integer example
age = 25
print(f"age = {age}, type = {type(age)}")

# Float example
price = 19.99
print(f"price = {price}, type = {type(price)}")

# Boolean example
is_active = True
print(f"is_active = {is_active}, type = {type(is_active)}")

# List example
colors = ["red", "blue", "green"]
print(f"colors = {colors}, type = {type(colors)}")
````

## 🔤 String (`str`)

A **string** is a sequence of characters (text) enclosed in quotes.

### Characteristics
- Created with **single quotes** `' '` or **double quotes** `" "`
- Can contain letters, numbers, symbols, and spaces
- **Immutable** (cannot be changed after creation)
- Support for **escape sequences** like `\n`, `\t`

### Examples
```python
name = "Alice"              # Double quotes
city = 'New York'           # Single quotes
sentence = "It's awesome"   # Mixed quotes when needed
```

### String Operations
- **Concatenation**: `"Hello" + " " + "World"`
- **Repetition**: `"Hi" * 3` → `"HiHiHi"`
- **Length**: `len("Python")` → `6`

````python
# String Examples

# Example 1: Creating strings
name = "Alice"
message = 'Hello World'
sentence = "It's a beautiful day"
print("Example 1 - Creating Strings:")
print(f"name = {name}")
print(f"message = {message}")
print(f"sentence = {sentence}\n")

# Example 2: String concatenation
first_name = "John"
last_name = "Doe"
full_name = first_name + " " + last_name
print("Example 2 - Concatenation:")
print(f"full_name = {full_name}\n")

# Example 3: String length
text = "Python"
print("Example 3 - String Length:")
print(f"len('{text}') = {len(text)}\n")

# Example 4: String repetition
repeated = "Hey" * 3
print("Example 4 - String Repetition:")
print(f"repeated = {repeated}")
````

## 🔢 Integer (`int`)

An **integer** is a whole number (positive, negative, or zero) **without decimals**.

### Characteristics
- **No decimal point**
- **Unlimited length** (can be very large)
- Can be **positive or negative**
- Support for **arithmetic operations** (+, -, *, /)

### Examples
```python
small = 42
large = 9876543210
negative = -123
zero = 0
```

### Integer Operations
- **Addition**: `5 + 3` → `8`
- **Subtraction**: `10 - 4` → `6`
- **Multiplication**: `3 * 7` → `21`
- **Division**: `20 / 4` → `5.0` (returns float!)
- **Floor Division**: `20 // 4` → `5` (returns int)
- **Modulo**: `17 % 5` → `2` (remainder)

````python
# Integer Examples

# Example 1: Creating integers
num1 = 42
num2 = -15
num3 = 0
print("Example 1 - Creating Integers:")
print(f"num1 = {num1}, type = {type(num1)}")
print(f"num2 = {num2}, type = {type(num2)}")
print(f"num3 = {num3}, type = {type(num3)}\n")

# Example 2: Large integers
large_num = 9876543210
print("Example 2 - Large Integer:")
print(f"large_num = {large_num}\n")

# Example 3: Arithmetic operations
a = 10
b = 3
print("Example 3 - Arithmetic Operations:")
print(f"{a} + {b} = {a + b}")
print(f"{a} - {b} = {a - b}")
print(f"{a} * {b} = {a * b}")
print(f"{a} / {b} = {a / b}")
print(f"{a} // {b} = {a // b} (floor division)")
print(f"{a} % {b} = {a % b} (modulo/remainder)")
````

## 🔢 Float (`float`)

A **float** is a number that contains one or more **decimal points**. Also called "floating point numbers".

### Characteristics
- **Contains decimal point**
- Can be **positive or negative**
- Used for **precise measurements** and **calculations**
- Default when dividing integers

### Examples
```python
price = 19.99
pi = 3.14159
negative = -5.5
small = 0.001
```

### Special Float Notations

#### Scientific Notation
Use **e** (or **E**) to indicate power of 10:
```python
large = 35e3        # 35 × 10³ = 35000.0
small = 1.5e-2      # 1.5 × 10⁻² = 0.015
```

### Float Operations
Same as integers: **+, -, *, /, //, %**

````python
# Float Examples

# Example 1: Creating floats
price = 19.99
pi = 3.14159
negative = -5.5
print("Example 1 - Creating Floats:")
print(f"price = {price}, type = {type(price)}")
print(f"pi = {pi}")
print(f"negative = {negative}\n")

# Example 2: Scientific notation
large_num = 35e3        # 35 × 10³
small_num = 1.5e-2      # 1.5 × 10⁻²
print("Example 2 - Scientific Notation:")
print(f"35e3 = {large_num}")
print(f"1.5e-2 = {small_num}\n")

# Example 3: Float arithmetic
x = 10.5
y = 3.2
print("Example 3 - Float Arithmetic:")
print(f"{x} + {y} = {x + y}")
print(f"{x} - {y} = {x - y}")
print(f"{x} * {y} = {x * y}")
print(f"{x} / {y} = {x / y}\n")

# Example 4: Division with integers returns float
a = 20
b = 4
result = a / b
print("Example 4 - Division Returns Float:")
print(f"{a} / {b} = {result}, type = {type(result)}")
````

## ✅ Boolean (`bool`)

A **boolean** is a data type with only **two possible values**: `True` or `False`.

### Characteristics
- **Only two values**: `True` or `False`
- **Capitalized** (always start with uppercase)
- Result of **comparisons** and **logical operations**
- Essential for **conditional logic** (if statements)

### Booleans from Comparisons
```python
5 > 3           # True
5 < 3           # False
5 == 5          # True
5 != 3          # True
```

### Booleans from Logical Operations
```python
True and False  # False
True or False   # True
not True        # False
```

### Truthy and Falsy Values
In Python, any value can be evaluated as **truthy** or **falsy**:
- **Truthy**: Non-zero numbers, non-empty strings, lists, etc.
- **Falsy**: 0, empty string `""`, empty list `[]`, `None`

````python
# Boolean Examples

# Example 1: Creating booleans
is_active = True
is_student = False
print("Example 1 - Creating Booleans:")
print(f"is_active = {is_active}, type = {type(is_active)}")
print(f"is_student = {is_student}, type = {type(is_student)}\n")

# Example 2: Booleans from comparisons
a = 10
b = 5
print("Example 2 - Comparisons:")
print(f"{a} > {b} = {a > b}")
print(f"{a} < {b} = {a < b}")
print(f"{a} == {b} = {a == b}")
print(f"{a} != {b} = {a != b}\n")

# Example 3: Logical operations
print("Example 3 - Logical Operations:")
print(f"True and False = {True and False}")
print(f"True or False = {True or False}")
print(f"not True = {not True}\n")

# Example 4: bool() function
print("Example 4 - bool() Function (Truthy/Falsy):")
print(f"bool(1) = {bool(1)}")
print(f"bool(0) = {bool(0)}")
print(f"bool('Hello') = {bool('Hello')}")
print(f"bool('') = {bool('')}")
print(f"bool([1, 2, 3]) = {bool([1, 2, 3])}")
print(f"bool([]) = {bool([])}")
````

## 📋 List (`list`)

A **list** is a **collection of values** stored in a **single variable**. Lists are **ordered and changeable**.

### Characteristics
- **Enclosed in square brackets** `[ ]`
- **Ordered** (maintains insertion order)
- **Changeable** (can modify, add, or delete items)
- **Allows duplicates**
- Can contain **mixed data types**

### Examples
```python
fruits = ["apple", "banana", "cherry"]
numbers = [1, 2, 3, 4, 5]
mixed = ["Alice", 25, 19.99, True]
empty = []
```

### List Operations

| Operation | Code | Result |
|-----------|------|--------|
| **Access item** | `fruits[0]` | `"apple"` |
| **List length** | `len(fruits)` | `3` |
| **Add item** | `fruits.append("orange")` | Adds to end |
| **Insert item** | `fruits.insert(1, "grape")` | Inserts at position |
| **Remove item** | `fruits.remove("apple")` | Removes value |

### Indexing
- **Positive index**: `0, 1, 2, ...` (start from beginning)
- **Negative index**: `-1, -2, -3, ...` (start from end)

````python
# List Examples

# Example 1: Creating lists
fruits = ["apple", "banana", "cherry"]
numbers = [1, 2, 3, 4, 5]
mixed = ["Alice", 25, 19.99, True]
print("Example 1 - Creating Lists:")
print(f"fruits = {fruits}")
print(f"numbers = {numbers}")
print(f"mixed = {mixed}\n")

# Example 2: Accessing list items
print("Example 2 - Accessing Items:")
print(f"First fruit = {fruits[0]}")
print(f"Last fruit = {fruits[-1]}")
print(f"Second number = {numbers[1]}\n")

# Example 3: List length
print("Example 3 - List Length:")
print(f"Length of fruits = {len(fruits)}")
print(f"Length of numbers = {len(numbers)}\n")

# Example 4: Modifying lists
colors = ["red", "green", "blue"]
print("Example 4 - Modifying Lists:")
print(f"Original: {colors}")
colors.append("yellow")
print(f"After append: {colors}")
colors[0] = "pink"
print(f"After changing first item: {colors}")
colors.remove("green")
print(f"After removing green: {colors}")
````

## 📦 Tuple (`tuple`)

A **tuple** is a **collection of values** similar to a list, but **ordered and unchangeable** (immutable).

### Characteristics
- **Enclosed in parentheses** `( )`
- **Ordered** (maintains insertion order)
- **Unchangeable** (cannot modify items after creation)
- **Allows duplicates**
- Can contain **mixed data types**
- Generally **faster** than lists

### Examples
```python
coordinates = (10, 20)
colors = ("red", "green", "blue")
mixed = ("Alice", 25, True)
single = (1,)        # Comma required for single item!
```

### Tuple Operations

| Operation | Code | Result |
|-----------|------|--------|
| **Access item** | `colors[0]` | `"red"` |
| **Tuple length** | `len(colors)` | `3` |
| **Check membership** | `"red" in colors` | `True` |

### Important Differences from Lists

| Feature | List | Tuple |
|---------|------|-------|
| **Brackets** | `[ ]` | `( )` |
| **Changeable** | ✅ Yes | ❌ No |
| **Speed** | Slower | Faster |
| **Use for** | Dynamic data | Fixed data |

````python
# Tuple Examples

# Example 1: Creating tuples
colors = ("red", "green", "blue")
coordinates = (10, 20)
mixed = ("Alice", 25, True)
print("Example 1 - Creating Tuples:")
print(f"colors = {colors}, type = {type(colors)}")
print(f"coordinates = {coordinates}")
print(f"mixed = {mixed}\n")

# Example 2: Accessing tuple items
print("Example 2 - Accessing Items:")
print(f"First color = {colors[0]}")
print(f"Last color = {colors[-1]}")
print(f"Second coordinate = {coordinates[1]}\n")

# Example 3: Tuple length
print("Example 3 - Tuple Length:")
print(f"Length of colors = {len(colors)}")
print(f"Length of coordinates = {len(coordinates)}\n")

# Example 4: Tuples are unchangeable
print("Example 4 - Tuples are Unchangeable:")
print(f"Original tuple: {colors}")
try:
    colors[0] = "yellow"  # This will cause an error
except TypeError as e:
    print(f"Error: Cannot modify tuple - {e}\n")

# Example 5: Checking membership
print("Example 5 - Checking Membership:")
print(f"'red' in colors = {'red' in colors}")
print(f"'yellow' in colors = {'yellow' in colors}")
````

## 🎲 Set (`set`)

A **set** is a **collection of unique values** that is **unordered and unchangeable**.

### Characteristics
- **Enclosed in curly braces** `{ }`
- **Unordered** (no specific order)
- **No duplicates** (automatically removes duplicates)
- **Changeable** (can add/remove items, but not modify)
- **Great for removing duplicates** from data

### Examples
```python
numbers = {1, 2, 3, 4, 5}
colors = {"red", "green", "blue"}
mixed = {1, "hello", 3.14, True}
empty = set()           # Empty set (not {})
duplicates = {1, 1, 2, 2, 3}  # Becomes {1, 2, 3}
```

### Set Operations

| Operation | Code | Result |
|-----------|------|--------|
| **Add item** | `colors.add("yellow")` | Adds new item |
| **Remove item** | `colors.remove("red")` | Removes item |
| **Set length** | `len(colors)` | Number of items |
| **Check membership** | `"red" in colors` | `True/False` |
| **Union** | `set1 \| set2` | Combines sets |
| **Intersection** | `set1 & set2` | Common elements |

### Important Characteristics

- **Cannot access by index** - `set[0]` will cause error!
- **Automatically removes duplicates**
- **Unordered** - each time you print, order might differ
- **Great for** membership testing and removing duplicates

````python
# Set Examples

# Example 1: Creating sets
colors = {"red", "green", "blue"}
numbers = {1, 2, 3, 4, 5}
mixed = {1, "hello", 3.14}
print("Example 1 - Creating Sets:")
print(f"colors = {colors}, type = {type(colors)}")
print(f"numbers = {numbers}\n")

# Example 2: Sets automatically remove duplicates
duplicates = {1, 1, 2, 2, 3, 3}
print("Example 2 - Automatic Duplicate Removal:")
print(f"Input: {{1, 1, 2, 2, 3, 3}}")
print(f"Result: {duplicates}\n")

# Example 3: Adding and removing items
items = {"apple", "banana", "cherry"}
print("Example 3 - Adding and Removing Items:")
print(f"Original: {items}")
items.add("orange")
print(f"After add: {items}")
items.remove("banana")
print(f"After remove: {items}\n")

# Example 4: Set length
print("Example 4 - Set Length:")
print(f"Number of colors = {len(colors)}\n")

# Example 5: Checking membership
print("Example 5 - Checking Membership:")
print(f"'red' in colors = {'red' in colors}")
print(f"'yellow' in colors = {'yellow' in colors}\n")

# Example 6: Set operations
set1 = {1, 2, 3}
set2 = {3, 4, 5}
print("Example 6 - Set Operations:")
print(f"set1 = {set1}")
print(f"set2 = {set2}")
print(f"Union (set1 | set2) = {set1 | set2}")
print(f"Intersection (set1 & set2) = {set1 & set2}")
````

## 🔑 Dictionary (`dict`)

A **dictionary** is a **collection of key-value pairs** that is **ordered and changeable**.

### Characteristics
- **Enclosed in curly braces** `{ }`
- **Key-value pairs**: Each item has a unique key and associated value
- **Ordered** (maintains insertion order - Python 3.7+)
- **Changeable** (can add, remove, or modify items)
- **No duplicate keys** (but values can be duplicated)
- **Keys must be immutable** (strings, numbers, tuples)

### Examples
```python
person = {"name": "Alice", "age": 25, "city": "NYC"}
student = {1: "John", 2: "Jane", 3: "Bob"}
mixed = {"name": "Alice", "age": 25, "scores": [95, 87, 92]}
```

### Dictionary Operations

| Operation | Code | Result |
|-----------|------|--------|
| **Access value** | `person["name"]` | `"Alice"` |
| **Add/Update** | `person["age"] = 26` | Updates age |
| **Remove** | `del person["age"]` or `.pop("age")` | Removes key-value |
| **Get keys** | `person.keys()` | All keys |
| **Get values** | `person.values()` | All values |
| **Get items** | `person.items()` | All key-value pairs |

### Why Use Dictionaries?

- **Named access**: Use meaningful keys instead of numeric indices
- **Better readability**: `person["name"]` is clearer than `person[0]`
- **Flexible data**: Store related information together

````python
# Dictionary Examples

# Example 1: Creating dictionaries
person = {"name": "Alice", "age": 25, "city": "New York"}
student = {1: "John", 2: "Jane", 3: "Bob"}
print("Example 1 - Creating Dictionaries:")
print(f"person = {person}")
print(f"student = {student}\n")

# Example 2: Accessing values
print("Example 2 - Accessing Values:")
print(f"Name: {person['name']}")
print(f"Age: {person['age']}")
print(f"Student 2: {student[2]}\n")

# Example 3: Adding and updating items
print("Example 3 - Adding and Updating Items:")
print(f"Original: {person}")
person["age"] = 26  # Update existing key
print(f"After update: {person}")
person["email"] = "alice@example.com"  # Add new key-value
print(f"After adding: {person}\n")

# Example 4: Removing items
car = {"brand": "Toyota", "model": "Camry", "year": 2022}
print("Example 4 - Removing Items:")
print(f"Original: {car}")
del car["year"]
print(f"After delete: {car}\n")

# Example 5: Getting dictionary info
info = {"python": "Language", "java": "Compiled", "javascript": "Web"}
print("Example 5 - Dictionary Operations:")
print(f"Keys: {list(info.keys())}")
print(f"Values: {list(info.values())}")
print(f"Items: {list(info.items())}")
print(f"Length: {len(info)}")
````

## 🔄 Type Conversion (Casting)

**Type conversion** means converting a value from one data type to another.

### Built-in Type Conversion Functions

| Function | Converts to | Example |
|----------|------------|---------|
| `int()` | Integer | `int("42")` → `42` |
| `float()` | Float | `float("3.14")` → `3.14` |
| `str()` | String | `str(123)` → `"123"` |
| `bool()` | Boolean | `bool(1)` → `True` |
| `list()` | List | `list("abc")` → `['a', 'b', 'c']` |

### Important Notes

- **String to Number**: String must contain valid number
  - ✅ `int("42")` works
  - ❌ `int("abc")` causes error

- **Float to Integer**: Decimal part is **truncated** (not rounded)
  - `int(3.99)` → `3` ❌ NOT `4`

- **Boolean to Number**:
  - `True` → `1`
  - `False` → `0`

````python
# Type Conversion Examples

# Example 1: String to Integer
print("Example 1 - String to Integer:")
str_num = "42"
int_num = int(str_num)
print(f"str_num = '{str_num}', type = {type(str_num)}")
print(f"int_num = {int_num}, type = {type(int_num)}\n")

# Example 2: String to Float
print("Example 2 - String to Float:")
str_decimal = "3.14"
float_num = float(str_decimal)
print(f"str_decimal = '{str_decimal}', type = {type(str_decimal)}")
print(f"float_num = {float_num}, type = {type(float_num)}\n")

# Example 3: Number to String
print("Example 3 - Number to String:")
num = 100
str_from_num = str(num)
print(f"num = {num}, type = {type(num)}")
print(f"str_from_num = '{str_from_num}', type = {type(str_from_num)}\n")

# Example 4: Float to Integer (Truncation)
print("Example 4 - Float to Integer (Truncation):")
float_val = 3.99
int_val = int(float_val)
print(f"float_val = {float_val}")
print(f"int_val = {int_val} (note: truncated, not rounded)\n")

# Example 5: Boolean Conversion
print("Example 5 - Boolean Conversion:")
print(f"bool(1) = {bool(1)}")
print(f"bool(0) = {bool(0)}")
print(f"bool('text') = {bool('text')}")
print(f"bool('') = {bool('')}")
````

### Important Reminders

📌 **Strings**: Enclosed in quotes
- ✅ `"Hello"` or `'Hello'`
- ❌ `Hello` (error)

📌 **Lists vs Tuples**: Very similar, but...
- Lists `[ ]` are **changeable** (mutable)
- Tuples `( )` are **unchangeable** (immutable)

📌 **Sets vs Dictionaries**: Both use curly braces, but...
- Sets `{1, 2, 3}` store unique values
- Dicts `{"a": 1}` store key-value pairs

📌 **Type Conversion Gotchas**
- `int("hello")` → Error (string not a number)
- `int(3.99)` → `3` (truncates, doesn't round)
- `str(123)` → `"123"` (now you can concatenate!)
- Sets automatically **remove duplicates**

### Common Operations by Data Type

#### **String Operations**
```python
len(str)                # Length
str + str              # Concatenation
str * num              # Repetition
```

#### **Number Operations**
```python
x + y, x - y, x * y    # Basic math
x / y                  # Division (returns float)
x // y                 # Floor division (returns int)
x % y                  # Modulo/Remainder
```

#### **List Operations**
```python
list[index]            # Access by index
len(list)              # Length
list.append(item)      # Add item
list.remove(item)      # Remove item
```

#### **Tuple Operations**
```python
tuple[index]           # Access by index
len(tuple)             # Length
# Cannot modify tuples!
```

#### **Set Operations**
```python
set.add(item)          # Add item
set.remove(item)       # Remove item
set1 | set2            # Union
set1 & set2            # Intersection
# No index access!
```

#### **Dictionary Operations**
```python
dict[key]              # Access by key
dict[key] = value      # Add/update
del dict[key]          # Remove
dict.keys()            # Get all keys
dict.values()          # Get all values
dict.items()           # Get all key-value pairs
```
