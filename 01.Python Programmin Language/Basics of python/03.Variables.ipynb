## 🎯 What are Variables?

**Variables** are **containers for storing data values** in Python. Think of them as labeled boxes where you can store information.

### Key Points
- Variables **do not need explicit type declaration** (Python infers the type)
- Variables can **change type after they are set**
- Variables make code **readable and maintainable**
- Essential for **storing data temporarily during program execution**

### Why Use Variables?
✅ Store data for later use  
✅ Make code more readable  
✅ Reuse values easily  
✅ Perform operations on stored data  

---

## 📋 Rules for Python Variable Names

Every programming language has naming rules. Follow these strictly:

| Rule | ✅ Valid | ❌ Invalid |
|------|---------|-----------|
| Must start with **letter or underscore** | `myVar` | `2myVar` |
| Can contain **letters, numbers, underscores** | `my_var_2` | `my-var` |
| **Case-sensitive** | `age` ≠ `Age` | Same variable twice |
| **Cannot be Python keywords** | `myif` | `if`, `for`, `while` |
| **No spaces allowed** | `my_var` | `my var` |

### Examples of Legal Variable Names
```python
myvar = "John"
my_var = "John"
_my_var = "John"
myVar = "John"          # camelCase
MYVAR = "John"
myvar2 = "John"
```

### Examples of Illegal Variable Names
```python
2myvar = "John"         # ❌ Starts with number
my-var = "John"         # ❌ Contains hyphen
my var = "John"         # ❌ Contains space
```

## ✨ Creating Variables

Python has **no explicit declaration statement** for variables. 

**A variable is created the moment you assign a value to it** using the `=` operator.

### Syntax
```python
variable_name = value
```

### Assignment Operator
- `=` assigns a value to a variable
- The variable name goes on the **left side**
- The value goes on the **right side**

### 3 Essential Steps
1. **Choose a valid name** (following the rules above)
2. **Use the assignment operator** `=`
3. **Assign a value** (number, string, list, etc.)

````python
# Creating Variables - Practical Examples

# Example 1: String Variable
name = "Alice"
print("Name:", name)

# Example 2: Integer Variable
age = 25
print("Age:", age)

# Example 3: Float Variable
height = 5.6
print("Height:", height)

# Example 4: List Variable
fruits = ["apple", "banana", "cherry"]
print("Fruits:", fruits)

# Example 5: Boolean Variable
is_student = True
print("Is Student:", is_student)
````

## 🔄 Data Types and Type Conversion (Casting)

### Common Data Types in Python

| Type | Example | Function |
|------|---------|----------|
| **String** | `"Hello"` | `str()` |
| **Integer** | `42` | `int()` |
| **Float** | `3.14` | `float()` |
| **Boolean** | `True` | `bool()` |
| **List** | `[1, 2, 3]` | `list()` |
| **Tuple** | `(1, 2, 3)` | `tuple()` |

### What is Type Conversion (Casting)?

**Type conversion** (or **casting**) is converting a value from one data type to another using built-in functions.

### When Do You Need Casting?

- Converting user input (always string) to a number
- Converting numbers to strings for display
- Converting between collection types
- Ensuring data is in the correct format for operations

````python
# Type Conversion (Casting) Examples - Foundation Level

# Example 1: Convert Integer to Float
num_int = 10
num_float = float(num_int)
print(f"Integer {num_int} becomes Float {num_float}")

# Example 2: Convert Float to Integer (Truncates decimal)
decimal = 3.99
whole_num = int(decimal)
print(f"Float {decimal} becomes Integer {whole_num}")

# Example 3: Convert String to Integer
str_number = "42"
number = int(str_number)
print(f"String '{str_number}' becomes Integer {number}")

# Example 4: Convert String to Float
str_decimal = "3.14"
decimal_num = float(str_decimal)
print(f"String '{str_decimal}' becomes Float {decimal_num}")

# Example 5: Convert Number to String
num = 100
str_num = str(num)
print(f"Number {num} becomes String '{str_num}'")
````

## 🔤 String Variables: Single or Double Quotes?

In Python, string variables can be declared using either **single quotes** `' '` or **double quotes** `" "`.

### Both are Equivalent
```python
name1 = "John"      # Double quotes
name2 = 'John'      # Single quotes
# Both create identical strings!
```

### When to Use Which?

| Situation | Use | Example |
|-----------|-----|---------|
| **Contains apostrophe** | Double quotes | `"It's working"` |
| **Contains double quote** | Single quotes | `'He said "Hi"'` |
| **When in doubt** | Double quotes | `"Hello, World"` |
| **Consistency** | Pick one style | Stick to your choice |

### Best Practice
Choose a style and **be consistent throughout your code**. Most Python projects prefer **double quotes**.

````python
# String Quotes Examples

# Example 1: Double Quotes
message1 = "Hello, World!"
print(message1)

# Example 2: Single Quotes
message2 = 'Hello, World!'
print(message2)

# Example 3: String with Apostrophe (use double quotes)
sentence1 = "It's a beautiful day"
print(sentence1)

# Example 4: String with Double Quote (use single quotes)
sentence2 = 'He said "Welcome!"'
print(sentence2)
````

## ⚠️ Variable Names are Case-Sensitive

In Python, **uppercase and lowercase letters are different**. 

### Important
- `age`, `Age`, and `AGE` are **three different variables**
- Each one stores its own separate value
- Python treats them as completely different names

### Example
```python
count = 5        # lowercase 'count'
Count = 10       # uppercase 'C'
COUNT = 15       # all uppercase

# All three are different variables!
```

### Best Practice
- Use **lowercase** for regular variables (Python convention)
- Use **UPPERCASE** for constants (values that don't change)
- Avoid confusion by choosing names wisely

````python
# Case-Sensitivity Examples

# Example 1: Different case = Different variables
age = 20
print(f"age = {age}")

Age = 30
print(f"Age = {Age}")

AGE = 40
print(f"AGE = {AGE}")

# Example 2: Modifying one doesn't affect others
age = age + 5
print(f"\nAfter modification:")
print(f"age = {age}")
print(f"Age = {Age}")    # Unchanged
print(f"AGE = {AGE}")    # Unchanged
````

## 📝 Variable Names with Multiple Words

Variable names with more than one word can be difficult to read. There are three popular naming conventions:

### 1️⃣ Camel Case (camelCase)
- First word starts with **lowercase**
- Each subsequent word starts with **uppercase**
- No separators used

```python
myVariableName = "John"
firstName = "Alice"
userAge = 25
```

### 2️⃣ Pascal Case (PascalCase)
- **Every word** starts with **uppercase**
- No separators used
- Often used for class names

```python
MyVariableName = "John"
FirstName = "Alice"
UserAge = 25
```

### 3️⃣ Snake Case (snake_case)
- All letters are **lowercase**
- Words separated by **underscores**
- **Python Convention** ⭐ (PEP 8 Style Guide)

```python
my_variable_name = "John"
first_name = "Alice"
user_age = 25
```

### 🎯 Recommendation
**Use snake_case** for Python variables (follows PEP 8 official Python style guide)

````python
# Naming Conventions Examples

# Camel Case
myVariableName = "John"
firstName = "Alice"
userAge = 25
print("Camel Case:")
print(f"myVariableName = {myVariableName}")

# Pascal Case
MyVariableName = "John"
FirstName = "Alice"
UserAge = 25
print("\nPascal Case:")
print(f"FirstName = {FirstName}")

# Snake Case (Python Convention)
my_variable_name = "John"
first_name = "Alice"
user_age = 25
print("\nSnake Case (Recommended for Python):")
print(f"first_name = {first_name}")
print(f"user_age = {user_age}")
````

## 🔄 Multiple Variable Assignment

### Assign Multiple Values to Multiple Variables

Python allows you to assign **different values to multiple variables in a single line**.

```python
x, y, z = "Orange", "Banana", "Cherry"
```

**Important:** The number of variables must **exactly match** the number of values!

### Assign One Value to Multiple Variables

You can also assign the **same value to multiple variables** in one line.

```python
x = y = z = "Orange"
```

All three variables now contain the same value.

````python
# Multiple Variable Assignment Examples

# Example 1: Multiple values to multiple variables
print("Example 1: Multiple values to multiple variables")
fruit1, fruit2, fruit3 = "Apple", "Banana", "Cherry"
print(f"fruit1 = {fruit1}")
print(f"fruit2 = {fruit2}")
print(f"fruit3 = {fruit3}")

# Example 2: One value to multiple variables
print("\nExample 2: One value to multiple variables")
x = y = z = "Orange"
print(f"x = {x}")
print(f"y = {y}")
print(f"z = {z}")

# Example 3: Multiple values with different types
print("\nExample 3: Multiple values with different types")
name, age, score = "Alice", 20, 95.5
print(f"Name: {name}, Age: {age}, Score: {score}")
````

## 📦 Unpacking Collections

If you have a **collection of values** in a list, tuple, etc., Python allows you to **extract the values into individual variables**. This is called **unpacking**.

### What is Unpacking?

Unpacking is extracting multiple values from a collection (list, tuple) and assigning them to **separate variables** in one line.

### Syntax
```python
variable1, variable2, variable3 = collection
```

### Important Rules
- Number of variables must **match** the number of items in the collection
- Works with lists, tuples, and other sequences
- Very helpful for extracting data efficiently

````python
# Unpacking Collections Examples

# Example 1: Unpack a list
print("Example 1: Unpacking a list")
fruits = ["Apple", "Banana", "Cherry"]
fruit1, fruit2, fruit3 = fruits
print(f"fruit1 = {fruit1}")
print(f"fruit2 = {fruit2}")
print(f"fruit3 = {fruit3}")

# Example 2: Unpack a tuple
print("\nExample 2: Unpacking a tuple")
colors = ("Red", "Green", "Blue")
color1, color2, color3 = colors
print(f"color1 = {color1}")
print(f"color2 = {color2}")
print(f"color3 = {color3}")

# Example 3: Unpack numbers
print("\nExample 3: Unpacking numbers")
numbers = [10, 20, 30]
a, b, c = numbers
print(f"Sum = {a + b + c}")
````

## 🖨️ Displaying Variables (Output)

The **`print()` function** is the primary way to **display variables** and their values.

### 3 Ways to Output Variables

#### **Method 1: Single Variable**
```python
x = "Python is awesome"
print(x)
```

#### **Method 2: Multiple Variables with Comma**
Separate multiple variables with **commas** inside `print()`:
```python
x = "Python"
y = "is"
z = "awesome"
print(x, y, z)              # Output: Python is awesome
```
- ✅ Works with **any data type**
- ✅ Automatically adds **spaces** between values
- ✅ **Recommended method** for multiple variables

#### **Method 3: String Concatenation**
Use the **`+` operator** to join strings:
```python
x = "Python "
y = "is "
z = "awesome"
print(x + y + z)            # Output: Python is awesome
```
- ⚠️ Only works with **strings**
- ⚠️ You must add spaces manually
- ⚠️ Cannot mix strings with numbers

#### **Method 4: F-Strings (Modern Python)**
```python
name = "Alice"
age = 25
print(f"Name: {name}, Age: {age}")
```
- ✅ Clean and readable
- ✅ Works with any data type
- ✅ Allows expressions inside braces

### Best Practice
Use **commas** for simplicity or **f-strings** for formatting.

````python
# Output Variables Examples

# Method 1: Single Variable
print("Method 1: Single Variable")
message = "Python is awesome"
print(message)

# Method 2: Multiple Variables with Comma
print("\nMethod 2: Multiple Variables (with comma)")
word1 = "Python"
word2 = "is"
word3 = "awesome"
print(word1, word2, word3)

# Method 3: String Concatenation
print("\nMethod 3: String Concatenation (with +)")
part1 = "Hello "
part2 = "World"
print(part1 + part2)

# Method 4: F-Strings (Modern & Recommended)
print("\nMethod 4: F-Strings (Recommended)")
name = "Alice"
age = 25
city = "New York"
print(f"Name: {name}, Age: {age}, City: {city}")

# Method 4b: F-Strings with expressions
print("\nMethod 4b: F-String with calculations")
num1 = 10
num2 = 20
print(f"Sum: {num1 + num2}")
print(f"Product: {num1 * num2}")
````
