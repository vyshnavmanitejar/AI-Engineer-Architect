What is Artificial Intelligence?
- Artificial intelligence (AI) is a branch of computer science that develops systems capable of performing tasks typically requiring human intelligence, such as reasoning, learning, problem-solving, and understanding language.

Augmented intelligence?
- Augmented intelligence is a human-centered partnership model where AI and machine learning enhance human cognitive performance, learning, and decision-making rather than replacing them.
- It serves as a tool for augmenting human capabilities—specifically by analyzing vast data—to improve accuracy in fields like healthcare, finance, and customer service.
- Key Aspects of Augmented Intelligence:
    - Human-in-the-Loop: Humans maintain control over the final decision, using AI to suggest options and provide insights.
    - Benefits: It enhances task completion speed, manages large datasets, removes human fatigue/bias from data interpretation, and provides better predictions.
    - Distinction from AI: While Artificial Intelligence (AI) often aims for automation and autonomy, augmented intelligence focus on augmenting human intelligence, sometimes referred to as intelligence amplification.
    - Applications: It is used in virtual assistants (Siri/Alexa), medical diagnostics support, personalized online shopping, and predictive analytics.

What is Machine Learning?
- Machine learning (ML) is a subset of artificial intelligence (AI) that develops algorithms and models allowing computers to learn patterns from data and make decisions or predictions without being explicitly programmed for every specific task.

What is Natural Language Processing?
- Natural language processing (NLP) is a subfield of computer science and artificial intelligence (AI) that uses machine learning to enable computers to understand and communicate with human language.

What is Deep Learning?
- Deep learning is a specialized subset of machine learning that uses multi-layered artificial neural networks to simulate human-like decision-making.

What is Generative AI?
- Generative Artificial Intelligence (GenAI) is a type of AI that creates new content—including text, images, code, audio, and video—by learning patterns from existing data. 
- Generative AI creates new content such as text, images, music, and videos from scratch using deep learning and large datasets.
- Unlike traditional AI that analyzes data, generative AI uses large language models (LLMs) to generate human-like text and perform tasks like translation and summarization.

- Capabilities of Generative AI

    - It can produce diverse content formats, engage in human-like conversations, and augment data to improve machine learning models.
    - Generative AI is projected to significantly boost global economic growth and productivity over the next decade.

- Use Cases Across Sectors

    - In marketing and creative fields, it personalizes ads and generates digital art, music, and videos.
    - In healthcare, gaming, fashion, and education, it supports tailored treatments, creates interactive game worlds, designs virtual try-ons, and customizes learning materials.


What are Large Language Models?
- Large Language Models (LLMs) is a subset of Generative AI and are advanced AI systems based on transformer architectures (e.g., GPT, Gemini) trained on massive datasets to understand, generate, and summarize human language.

What is Agentic AI?
- Agentic AI refers to autonomous systems that go beyond generating content to actively achieving complex, multi-step goals with minimal human oversight.
- These AI agents plan, reason, use external tools, and adapt their actions in real time. Unlike passive generative AI, they transform knowledge into proactive, autonomous action.

What is AI Agents?
- An AI agent is a software program that acts autonomously to achieve specific goals, moving beyond simple text generation to perform multi-step tasks, reason, plan, and use external tools.
- Powered by Large Language Models (LLMs), these agents perceive their environment, learn from data, and interact with applications (e.g., browsing, sending emails) with minimal human oversight.
