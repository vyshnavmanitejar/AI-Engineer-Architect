```python
import pandas as pd
import numpy as np

def inspect_data(df, target_col=None):
    print("=" * 80)
    print("1) SHAPE")
    print("=" * 80)
    print("Rows, Columns:", df.shape)

    print("\n" + "=" * 80)
    print("2) SAMPLE ROWS")
    print("=" * 80)
    print(df.head())

    print("\n" + "=" * 80)
    print("3) COLUMN NAMES")
    print("=" * 80)
    print(df.columns.tolist())

    print("\n" + "=" * 80)
    print("4) DATA TYPES / INFO")
    print("=" * 80)
    df.info()

    print("\n" + "=" * 80)
    print("5) MISSING VALUES")
    print("=" * 80)
    missing = df.isnull().sum().sort_values(ascending=False)
    missing_pct = ((df.isnull().sum() / len(df)) * 100).sort_values(ascending=False)
    missing_report = pd.DataFrame({
        "missing_count": missing,
        "missing_percent": missing_pct.round(2)
    })
    print(missing_report[missing_report["missing_count"] > 0])

    print("\n" + "=" * 80)
    print("6) DUPLICATES")
    print("=" * 80)
    print("Duplicate rows:", df.duplicated().sum())

    print("\n" + "=" * 80)
    print("7) UNIQUE VALUES PER COLUMN")
    print("=" * 80)
    print(df.nunique().sort_values())

    print("\n" + "=" * 80)
    print("8) NUMERICAL SUMMARY")
    print("=" * 80)
    num_cols = df.select_dtypes(include=np.number).columns
    if len(num_cols) > 0:
        print(df[num_cols].describe().T)
    else:
        print("No numerical columns found.")

    print("\n" + "=" * 80)
    print("9) CATEGORICAL SUMMARY")
    print("=" * 80)
    cat_cols = df.select_dtypes(include=["object", "category", "bool"]).columns
    if len(cat_cols) > 0:
        print(df[cat_cols].describe().T)
    else:
        print("No categorical columns found.")

    print("\n" + "=" * 80)
    print("10) DATE COLUMNS CHECK")
    print("=" * 80)
    print(df.select_dtypes(include=["datetime64[ns]", "datetime64"]).columns.tolist())

    print("\n" + "=" * 80)
    print("11) LOW-CARDINALITY COLUMNS")
    print("=" * 80)
    low_card = [col for col in df.columns if df[col].nunique() <= 10]
    for col in low_card:
        print(f"\nColumn: {col}")
        print(df[col].value_counts(dropna=False).head(10))

    if target_col is not None and target_col in df.columns:
        print("\n" + "=" * 80)
        print("12) TARGET COLUMN CHECK")
        print("=" * 80)
        print("Target column:", target_col)
        print("Missing in target:", df[target_col].isnull().sum())
        print("Unique target values:", df[target_col].nunique())
        print("\nTarget distribution:")
        print(df[target_col].value_counts(dropna=False))

    print("\n" + "=" * 80)
    print("13) SUSPICIOUS CHECKS")
    print("=" * 80)
    print("Columns with all null values:")
    print(df.columns[df.isnull().all()].tolist())

    print("\nColumns with only one unique value:")
    one_value_cols = [col for col in df.columns if df[col].nunique(dropna=False) <= 1]
    print(one_value_cols)

    print("\nDone: first-pass inspection complete.")


# Example usage
df = pd.read_csv("/Users/vyshnav/Downloads/Original_data_with_more_rows.csv")
inspect_data(df, target_col=None)
```
