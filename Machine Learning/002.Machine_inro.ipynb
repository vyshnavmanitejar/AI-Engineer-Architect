What is Machine Learning?
- Machine learning (ML) is a subset of artificial intelligence (AI) that develops algorithms and models allowing computers to learn patterns from data and make decisions or predictions without being explicitly programmed for every specific task.

Key components:
- Data
- Models (algorithms)
- Learning process
- Predictions

Types of machine learning methods:
- Supervised Learning method:
    - Supervised learning is a machine learning technique where models are trained using input, output, or x to y mapping and predict outputs for new, unseen data.
    - Technique: Regression, and Classification
    - Model: Linear regression, logistic regression
- Unsupervised Learning method:
    - Unsupervised learning is a type of machine learning technique that trains models on row, unlabeled datasets to automatically discover hidden patterns, structures, or relationships without human guidance.
    - Techinique: clustering
    - Models: Anomaly detection, dimensionality reduction.
- Reinforcement Learning method:
    - Reinforcement Learning (RL) is a branch of machine learning that focuses on how agents can learn to make decisions through trial and error to maximize cumulative rewards.
    - RL allows machines to learn by interacting with an environment and receiving feedback based on their actions. This feedback comes in the form of rewards or penalties.
    - Techinuques: value-based (Q-learning), policy-based (Policy Gradients), and hybrid (Actor-Critic)
    - Model: AI agents, Q-Learning, SARSA, and DQN
