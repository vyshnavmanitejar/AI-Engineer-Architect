## Core Architectures

### Variational Autoencoders (VAE)

- Summary: Probabilistic encoder–decoder that learns a continuous latent space.
- How it works: encoder → latent distribution (mean, variance) → sample → decoder; optimized by reconstruction loss + KL divergence.
- Strengths: smooth latent interpolation, principled probabilistic formulation, useful for representation learning and anomaly detection.
- Limitations: often produces blurrier images vs. adversarial approaches; trade-off between reconstruction fidelity and latent regularization.
- Common applications: image generation, data augmentation, anomaly detection, dimensionality reduction.

---

### Generative Adversarial Networks (GAN)

- Summary: Two-network adversarial setup (generator vs discriminator) that yields high-fidelity samples.
- How it works: generator produces samples; discriminator distinguishes real vs fake; adversarial loss drives generator improvements.
- Strengths: exceptional sample realism, strong empirical results for images (StyleGAN, BigGAN).
- Limitations: training instability, mode collapse, sensitive hyperparameters and architecture choices.
- Common applications: photo-realistic image synthesis, style transfer, super-resolution, dataset synthesis.

---

### Autoregressive Models

- Summary: Sequence models that predict next token conditioned on previous tokens (P(x_t | x_1..x_{t-1})).
- How it works: generate sequentially; each step conditions on past outputs (e.g., RNNs, Transformer decoder-only/GPT).
- Strengths: strong likelihood modeling, excellent for coherent text and sequence generation.
- Limitations: sequential generation can be slower at inference; long-range dependencies can require transformer variants.
- Common applications: language modeling, text generation, music, code generation, speech synthesis.

---

### Transformers (Modern Standard)

- Summary: Attention-based architectures that process sequences in parallel and scale effectively.
- How it works: self-attention computes context-aware representations; encoder/decoder or decoder-only variants enable understanding and/or generation.
- Strengths: parallelism (fast training), long-context handling, strong transfer learning behavior via pretraining + fine-tuning.
- Limitations: high compute/memory cost for very long contexts; many engineering optimizations required for production-scale models.
- Common applications: large language models (GPT family), multimodal models, machine translation, summarization, and vision transformers (ViT).

## Modalities: Unimodal vs Multimodal

- Unimodal: Models trained and evaluated within a single data modality (text→text, image→image, audio→audio). Examples: GPT-3 (text), StyleGAN (images), WaveNet (audio).
- Multimodal: Models that accept and/or produce multiple modalities (text+image, audio+text). Examples: DALL·E, GPT-4V, CLIP, ImageBind.

Advantages of multimodal systems include richer context, improved grounding, and broader application scope; they are more complex and require aligned datasets.

## Practical Notes & Evaluation

- Training tips: stabilize GANs with spectral normalization, progressive growing, two-time-scale updates; for transformers use learning-rate scheduling and large-batch strategies.
- Evaluation metrics:
  - Images: FID, IS (measure realism & diversity)
  - Text: BLEU/ROUGE (surface), BERTScore (semantic), human evaluation for coherence/factuality
  - Multimodal: task-specific retrieval/QA metrics + human evaluation
- Dataset quality: diverse, high-quality, and well-curated datasets reduce bias and improve sample realism.
- Compute considerations: pretraining LLMs and large diffusion/transformer models requires significant GPU/TPU resources and cost planning.

---

## Ethics, Safety & Deployment

- Hallucinations & factuality: LLMs can produce plausible but incorrect outputs—use retrieval-augmentation, grounding, and verification.
- Bias & fairness: monitor and mitigate biases originating from training data; apply fairness testing and dataset audits.
- Copyright & provenance: ensure training data licensing compliance and include provenance/attribution where needed.
- Misuse prevention: rate limits, content filters, moderation, and usage policies to reduce harmful outputs.
- Privacy: avoid training on sensitive personal data; apply differential privacy or data minimization when required.

---

## Key Takeaways

- Transformers are the dominant, scalable paradigm for modern generative models, especially for text and multimodal tasks.
- GANs excel at photorealistic image synthesis but require careful engineering to stabilize training.
- VAEs provide useful latent representations and probabilistic modeling, though sample fidelity may lag behind GANs.
- Model selection should match modality, fidelity, compute budget, and safety constraints.
