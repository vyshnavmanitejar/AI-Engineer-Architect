```

TRAINING PHASE           DEPLOYMENT PHASE         USE PHASE```

Data Security         Ownership Ambiguity    Bias Impacts

Data Collection   →     Model Creation    →    User AccessBias Embedding        IP Rights             Misuse/Deepfakes

  ↑                        ↑                       ↑Privacy Issues         Copyright Questions    Hallucinations

# ⚖️ Copyright Complexity in Generative AI: The Ownership Question

## The Central Question

**When an AI creates something, who owns it?**

This seemingly simple question has no clear answer under current law, creating a legal gray zone that affects billions in potential value.

---

## The Ownership Ambiguity Problem


### Who Could Own AI-Generated Content?



POSSIBLE CLAIMANTS:

1. **AI DEVELOPERS** - Built the system, created algorithm (vs. Did not directly create specific output)✨ **The copyright question in generative AI is unsettled, but movement is happening**

2. **THE USER/OPERATOR** - Gave the prompt/input, directed creation (vs. Minimal creative input)

3. **DATA HOLDERS** - Provided training data, AI learned from their content (vs. Training data licensed/public)☐ Cross-border consistency

4. **NOBODY (Public Domain)** - AI isn't conscious/creative, no human authorship (vs. Eliminates incentives)☐ Liability assignment

5. **EVERYONE (Collective)** - Society trained the AI, multiple stakeholders contributed (vs. Impractical)☐ Copyrightability standards for AI output

☐ Fair use standards for training data

**The reality: Different jurisdictions are answering this differently**☐ Clear ownership frameworks



---## Clear Guidelines: What We Need



## Traditional IP Framework vs. AI Reality---



TRADITIONAL: Human author → applies skill → creates original work → copyright granted- User agreement determines final ownership

- Developer owns rights initially

AI CREATION: User provides prompt → AI applies patterns → AI generates output → Who is "author"? What rights exist?- AI-generated works can be copyrighted

- More permissive toward AI development

**The problem:** Copyright requires "human authorship" under most laws### China



---- Developers liable for copyright violations

- Creators can opt-out of training data use

## Real-World Copyright Conflicts- Requires transparency about AI-generated content

- EU AI Act provides framework

**Getty Images vs. Stability AI:** Sued for using copyrighted photos in training, concerns about AI generating infringing images### European Union



**Authors vs. OpenAI:** Major authors sued for training on millions of copyrighted books without permission- Exception: If human substantially directed AI as a tool

- Purely AI-generated work NOT copyrightable

**US Copyright Office (2023):** AI-generated content may not be copyrightable—requires human authorship- Human creativity required for copyright

### United States

---

## Current State by Jurisdiction

# 🔒 Privacy in Generative AI: The Data Dilemma

## The Fundamental Problem

**Generative AI requires massive data to function effectively, but massive data + privacy = contradiction**

---

## Why Privacy Matters in Generative AI

### What Ends Up in Training Data?
- **Public data:** Books, articles, websites, academic papers, code

- **Personal data:** Medical records, financial info, social media posts, private messages

- **Confidential data:** Corporate secrets, legal documents, government records, trade secrets



**Result: AI models may contain pieces of sensitive information**✨ **Privacy-by-design principles must be part of development from day one, not added later**



### Privacy Threats**Privacy in generative AI is achievable but requires intentional design:**

1. **Memorization** - AI can memorize training data verbatim and retrieve with clever prompting

2. **Pattern Inference** - Even if not memorized, patterns leak information## Key Takeaway

3. **Misuse by Third Parties** - Model access sold to unintended recipients

4. **Regulatory Violations** - GDPR, CCPA requirements not met---



---- May be less capable than public models

- Less data available (only organization's data)

## Privacy by Design in Generative AI- Higher cost to build/maintain

### Trade-offs

### Strategy 1: Data Minimization

Only collect what you need → Remove PII → De-identify → Aggregate where possible- Competitive advantage preserved

- Regulatory compliance easier

### Strategy 2: Differential Privacy- Data stays internal and proprietary

Add mathematical noise to protect individuals while keeping statistics accurate### Benefits



### Strategy 3: Federated Learning**Private AI = AI system exclusive to one organization where data never leaves organizational control**

Train locally on devices → Only send updates (not raw data) to central server

## Private AI: Organizational Protection

### Strategy 4: Encryption & Access Control

✓ Encrypt at rest, in transit, and in use---

✓ Strict access controls

✓ Audit logs of all access✓ Data deletion mechanisms

# 🏢 Implementing Private AI: A Practical Framework

## 3-Layered Implementation

### Layer 1: Infrastructure Isolation

**Option A: On-Premises**
- Complete data control
- No data leaves premises
- Fastest performance
- High capital costs


**Option B: Hybrid (Private Cloud)**

- Cloud scalability

- No shared infrastructure

- Provider manages hardware✨ **The investment in privacy-first AI pays off in customer trust, regulatory compliance, and competitive advantage**

- Lower capital costs than on-prem

- ⚖️ Ongoing compliance

### Layer 2: Data Security- 📋 Clear governance

- Encryption at rest- 🔐 Comprehensive security

- Encryption in transit- 🏢 Proper infrastructure setup

- Encryption in use (advanced)**Private AI requires commitment to:**

- Access control & authentication

- Data minimization## Key Takeaway



### Layer 3: Operational Security---

- Policies (data classification, retention, incident response)

- Training (employee awareness)PHASE 1: Assessment → PHASE 2: Planning → PHASE 3: Implementation → PHASE 4: Testing → PHASE 5: Launch

- Monitoring (access logs, audits, alerts)

- Incident response (detection, containment, recovery)## Implementation Roadmap



------



## Custom Models: Tailored AI**Industry-Specific:** HIPAA (healthcare), FINRA (finance), FERPA (education)



### Approach 1: Fine-Tuning Public Models**CCPA Compliance:** Clear data sources (internal only), transparency easier, can opt out from specific uses

- Start with pre-trained model

- Fine-tune on organization's data**GDPR Compliance:** Data doesn't leave organization, easier to provide access, can comply with right to be forgotten

- Deploy privately

- **Advantages:** Faster, requires less data, cost-effective## Compliance-First Implementation

- **Drawbacks:** Inherits biases from public model

---

### Approach 2: Training from Scratch

- Gather private data- **Drawbacks:** Expensive, requires large datasets, significant expertise

- Train custom model- **Advantages:** Complete control, customized, no inherited biases

- Optimize & test- Deploy on internal infrastructure

# 🤖 AI Hallucinations: When Generative AI Makes Things Up

## What Is an AI Hallucination?

**AI Hallucination** = When an AI system confidently generates false, fabricated, or inaccurate information as if it were true.

### Key Characteristic
**The AI doesn't say: "I'm not sure" or "I don't know"**
**Instead: It states false information confidently as fact**

---

## Real-World Examples



**Fake Legal Citations:** Lawyer uses ChatGPT to find case law. AI cites cases that don't exist. Lawyer includes in legal brief. Judge discovers fabrication. Lawyer faces sanctions.



**Fake Academic Papers:** AI generates citations to non-existent research. Student uses in paper. Academic integrity violation results.✨ **Verification is essential for factual claims**

✨ **Confidence level is unrelated to accuracy**

**Inaccurate Product Info:** Business uses ChatGPT for customer service. AI makes up features product doesn't have. Customer buys expecting feature that doesn't exist.✨ **AI models play probability games, not retrieve facts**



---**AI Hallucinations are not a bug—they're a feature of how language models work:**



## Why Do Hallucinations Happen?## Key Takeaway



### Root Cause 1: No True Understanding---

- AI recognizes patterns, doesn't "know" facts

- AI predicts next most likely word- Use AI for draft, human for refinement

- AI doesn't verify against reality- Don't trust without verification

- AI prioritizes plausibility over accuracy- Verify citations independently

- Cross-check important facts

### Root Cause 2: Training Data Gaps**For Users:**

- Missing information filled with plausible-sounding content

- Contradictory data combined incorrectly- Train to express uncertainty

- Fact-checking layers

### Root Cause 3: Overconfidence- Retrieval-augmented generation (look up facts)

- Training penalizes "I don't know" responses- Better training data (higher quality, recent, domain-specific)

- Training rewards fluent, confident answers**For Developers:**

- AI learns: Always answer, sound confident

## Strategies to Reduce Hallucinations

---

---

## Types of Hallucinations

1. **Factual** - False facts stated as true4. **Domain Mismatch** - Users don't know where AI fails

2. **Logical** - Invalid reasoning presented3. **Subtle Errors** - Off-by-one errors hard to catch

3. **Attribution** - False quotes or citations2. **Scale and Speed** - Thousands of plausible-sounding outputs instantly

4. **Emergent** - Entirely fictional but plausible narratives1. **Confidence Without Basis** - Sounds credible, might be completely false



---## Why Hallucinations Are Dangerous


# 👺 Deepfakes: When AI Creates Convincing Lies

## What Are Deepfakes?

**Deepfakes** = Synthetic media (video, audio, images) showing someone doing or saying something they didn't actually do/say.

### Three Categories
1. **Replacement** - Replace Face A with Face B in video
2. **Audio** - Generate voice that sounds like specific person
3. **Hybrid** - Modify real video to change context/speech

---



## The Ethical Challenges



### Challenge 1: Misinformation at Scale✨ **Individuals must develop critical media consumption habits in the deepfake era**

- Hyper-realistic video "evidence"✨ **We need multi-layered approaches: technology, policy, platform governance, and media literacy**

- Spreads faster than debunking

- Fact-checkers lag behind**Deepfakes represent a fundamental challenge to trust:**

- Damage is massive and lasting

- **Example:** Deepfake video of politician released 1 week before election, spreads in hours, election already happened## Key Takeaway



### Challenge 2: Personal Violation---

- Non-consensual deepfakes, often sexual

- Primarily targets women- Reduce reach of misleading deepfakes

- PTSD, anxiety, depression for victims- Add watermarks/metadata

- Permanent internet record- Remove non-consensual intimate content

- Career damage- Flag deepfakes as synthetic

### Platform Responses

### Challenge 3: Impersonation & Fraud

- CEO impersonation for wire fraud- Enforcement difficult

- Voice cloning for "grandparent scams"- Proof of creation is hard

- Facial recognition bypass- Some US states: Criminalizing non-consensual deepfakes

- Real financial/security consequences- EU: Deepfakes classified as harmful

### Legal Approaches

### Challenge 4: Harassment & Blackmail

- Create explicit deepfake to humiliate- Detection lags behind generation (arms race)

- Used for cyberbullying- Audio analysis (frequency patterns)

- "Pay or I'll share it" extortion- Facial analysis (eye movements, blinks)

- Targets activists and minorities### Technical Defenses



### Challenge 5: Trust Erosion ("Liar's Dividend")## Addressing the Deepfake Challenge

- **Real evidence dismissed:** "This is a deepfake!"

- **Fake evidence believed:** "Could be real..."---

- Both real and fake evidence faces skepticism

- Democratic institutions undermined2016: Crude → 2018: Realistic → 2020: High-quality → 2024: Indistinguishable from real

### Quality Evolution

---

- **Accessibility:** Apps for smartphones, open-source code, no technical expertise needed

## Deepfake Technology- Neural Vocoders for audio

- VAEs (Variational Autoencoders)

### How They're Made- GANs (Generative Adversarial Networks)

# 📜 Intellectual Property in AI-Generated Content

## What Is Intellectual Property?

**IP** = Legal rights over creations of the mind (inventions, artistic works, symbols, names, images, designs)

### Traditional IP Categories
1. **Copyright** - Literary works, music, art, software (Life + 70 years)
2. **Patents** - Inventions, processes, new technology (20 years)
3. **Trademarks** - Brand names, logos, symbols (Renewable indefinitely)
4. **Trade Secrets** - Proprietary information with value (As long as kept secret)



---



## The Core Problem✨ **Get explicit written agreements about ownership before using AI commercially**

✨ **Most AI output: Public domain (no protection)**

**Traditional IP assumes: Human author + skill + creativity = protection**✨ **Rules differ by jurisdiction**

✨ **Current legal framework uncertain**

**AI-Generated Content: No human author, AI applied patterns, deserves protection?**

**IP for AI-generated content is in flux:**

---

## Key Takeaway

## Ownership Question

---

**Who might own AI-generated content?**

1. AI Developer - Created the system (but not output)**For Users:** Read terms of service, get written ownership agreements, review for copyright infringement

2. User/Prompts Giver - Directed creation (but minimal effort)

3. Data Contributors - AI learned from their work (but licensed/public)**For AI Companies:** Patent algorithms, keep models as trade secrets, licensing agreements

4. Nobody (Public Domain) - No human authorship (but no incentive)

5. Everyone (Collective) - Multiple stakeholders (but impractical)**For Content Creators:** Use anti-AI licenses, watermarking, opt-out clauses



---## Practical IP Strategies



## Current State by Jurisdiction---



**United States:** AI-generated content NOT copyrightable (requires human authorship)**China:** More permissive—AI-generated works CAN be copyrighted, promotes innovation

- Exception: If human substantially modified/selected AI output

**European Union:** EU AI Act requires transparency, creators can opt-out of training

# 👁️ Ethical Challenges: Hallucinations & Biases Combined

## The Double Threat

**Problem 1: AI Hallucinations**
- AI generates false information confidently
- Users can't tell what's real vs. fabricated
- Misinformation spreads at scale

**Problem 2: AI Biases**
- AI reflects biases from training data
- AI treats some groups unfairly

- Discrimination automated at scale



**Combined: Confidently biased misinformation amplified at scale**

✨ **Addressing both hallucinations and biases together is critical for ethical AI**

---

**Hallucinations + Biases = Amplified Harm:**

## Real-World Harms

## Key Takeaway

### Harm 1: Discrimination Through Misinformation

- Deepfake video shows minority person saying offensive thing---

- Video looks authentic, spreads on social media

- Person loses job, receives death threats- Seek human review for significant decisions

- From AI-generated lie- Question high-confidence outputs

- Verify important information

### Harm 2: Healthcare Disparities- Don't rely solely on AI

- AI trained on majority population data**For Users:**

- Minority patients present differently

- AI systematically underestimates severity- Transparency about limitations

- Wrong treatment recommended- Regular fairness audits

- Keep humans in the loop

### Harm 3: Criminal Justice- Human review for critical decisions

- Bail prediction AI with biased training data**For Organizations:**

- AI perpetuates historical discrimination

- Defendant denied bail, sits in jail- Train to express uncertainty

- Harmed by AI bias + confident hallucination- Fairness audits

- Diverse training data

---- Retrieval-augmented generation for facts

**For Developers:**

## Accountability Crisis

## Mitigation Strategies

**When things go wrong:**

- User denied loan asks: Who's responsible?---

- **Unclear:** Developer? Bank? Data scientist? No one?

- **Result:** Victim has no recourse→ Even good AI gets rejected, dysfunction spreads



---Initial Trust → Discovery of Problems → Generalized Distrust → Rejection



## Why Detection Is Hard## Trust Degradation Spiral



**Hallucinations:** Plausible but wrong, subtle errors, context-dependent---


**Biases:** Invisible in aggregate, blamed on data, happen through proxies

# 🤝 Ethical Guidelines for Developer-Built Fair AI

## 5-Pillar Fairness Framework

### Pillar 1: Identify and Mitigate Bias

**Step 1: Bias Audit**
- What groups are represented in training data?
- What groups are missing?
- Are representations proportional?
- What historical biases exist?

**Step 2: Bias Mitigation**

- Data level: Balance training data, oversample underrepresented groups

- Algorithm level: Add fairness constraints, adjust thresholds

- Post-processing: Calibrate predictions, implement fairness-aware ranking

✨ **Fair AI doesn't happen accidentally—it requires intentional design, ongoing monitoring, and genuine commitment to equity**

**Step 3: Continuous Monitoring**

- Weekly: Group-level accuracy metrics**Building fair AI requires commitment and discipline:**

- Monthly: Full fairness audit

- Quarterly: Comprehensive review## Key Takeaway



------



### Pillar 2: Ensure TransparencyPhase 1: Planning → Phase 2: Development → Phase 3: Testing → Phase 4: Deployment → Phase 5: Monitoring



**What Users Need to Know:**## Implementation Roadmap

1. **Why:** Decision explanation with factors

2. **Factors Used:** Which factors mattered and their weights---

3. **How to Appeal:** Appeal process and timeframe

4. **Data Used:** Where data came from**Frequency:** Weekly (critical), monthly (high), quarterly (medium)

5. **Limitations:** What system doesn't consider

- Harm assessment (real-world impact)

**Model Card Documentation:** Purpose, data, performance, fairness, risks, governance- Calibration (predicted probabilities accurate per group)

- Disparate impact (4/5 rule)

---- Equalized odds (error rates similar across groups)

- Demographic parity (approval rates same across groups)

### Pillar 3: Promote Accountability**Test Suite:**



**Clear Responsibility Assignment:** Use RACI matrix### Pillar 5: Test for Fairness Regularly

- R = Responsible | A = Accountable | C = Consulted | I = Informed

---

**Incident Response Plan:**

1. Assess severity (1 hour)- Subject matter experts

2. Immediate response (4 hours)- Legal/compliance

3. Investigate (24-72 hours)- Ethics experts

4. Remediate (1-2 weeks)- Affected communities

5. Learn (ongoing)**Stakeholder Engagement:**



---**Diverse Team:** Technical, domain, demographic, perspective diversity


### Pillar 4: Engage Diverse Perspectives
