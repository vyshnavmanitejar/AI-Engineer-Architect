``````

Loan Rejected  Contact: [governance team email]

Reason: "Credit score (680) below threshold (700)"  Legal: Complies with Fair Lending Act

Recourse: "Improve credit score or appeal"  Ethics: Approved by ethics committee

```REVIEW



### 3. **Clarifies Responsibility**  Mitigation: Manual review for flagged cases

  Risk: May disadvantage recent immigrants

When errors occur, transparency identifies **who** is accountable:KNOWN ISSUES

- **Developer:** Built wrong algorithm

- **Data team:** Used biased training data  Threshold: Score > 700 = Approved

- **Deployer:** Misconfigured thresholds  Factors: Credit score (40%), Income (30%), Debt ratio (20%), Employment history (10%)

- **Manager:** Failed to monitor performanceDECISION LOGIC



### 4. **Detects and Prevents Bias**  Approval rate: 68%

  Fairness: 2.1% demographic disparity

```  Accuracy: 92%

BIAS DETECTION PROCESSPERFORMANCE



1. Transparent Decision Logs  Transparency: See bias audit results

   ↓ What data was used? How was decision made?  Demographics: 45% Male, 55% Female

  Size: 100,000 applications

2. Audit for Discrimination  Source: Bank loan database (2020-2023)

   ↓ Are certain groups treated unfairly?DATA



3. Identify Root Cause  Deployed: Jan 2024

   ↓ Which factors caused the bias?  Owner: Financial Services AI Team

  Purpose: Pre-approval screening for loan applications

4. Implement FixOVERVIEW

   ↓ Retrain model, adjust thresholds, etc.

MODEL CARD: Loan Approval AI v2.1

5. Monitor Going Forward```

```

### Model Cards: Documenting AI Systems

---

## Transparency in Practice

## Types of Transparency

---

| Type | What It Means | Example |

|---|---|---|| **Impact Transparency** | Knowing real-world effects | "This system has 3.2% demographic disparity" |

| **Model Transparency** | Understanding how the algorithm works | "Decision tree shows: Age > 25 AND Score > 700" || **Outcome Transparency** | Explaining specific decisions | "Your application scored 720, approved" |

| **Data Transparency** | Knowing what data was used | "Trained on 50,000 loans from 2020-2023" || **Process Transparency** | Understanding the workflow | "Application → Scoring → Review → Decision" |

# 🔒 Data Privacy and Security in AI: Protecting Personal Information

## What Is Data Privacy in AI?

**Data privacy** means protecting individuals' personal information from unauthorized access, use, or disclosure—ensuring AI systems respect human dignity and rights.

**Data security** means implementing technical and organizational measures to prevent breaches and misuse.

### Why It's Different in AI

```

TRADITIONAL SOFTWARE        AI SYSTEMS



Data stored                 Data LEARNED

Accessible to few           Embedded in model✨ **Strong privacy protections enable responsible, trustworthy AI**

Can be replaced             Hard to "unlearn"

Breach = lost data          Breach = reverse engineer model**Privacy isn't a luxury—it's a fundamental human right:**

```

## Key Takeaway

---

---

## Core Privacy Principles

- Equitable innovation

### 1. **Minimize Data Collection**- Democratic safeguards against surveillance

- Protection of vulnerable populations

```- Reduced data exploitation

Collect Only What's Needed✅ **For Society:**



✗ Excessive: Age, Gender, Race, Religion, Sexual Orientation- Reduced breach costs

             Credit History, Medical Records, Political Affiliation- Brand reputation protection

- User trust and loyalty

✓ Justified: Income level, Employment, Credit score- Regulatory compliance

             (directly relevant to loan decision)✅ **For Organizations:**

```

- Control over their own data

### 2. **Get Informed Consent**- Reduced risk of discrimination or fraud

- Dignity and autonomy respected

**Users Should Know:**- Personal information protected

- What data you're collecting✅ **For Users:**

- How it will be used

- Who has access## Benefits of Privacy Protection

- How long it's kept

- Their rights (access, deletion, etc.)---



### 3. **Anonymize When Possible**```

10. Are we compliant with laws?

```9. What are user rights?

PERSONAL DATA                      ANONYMIZED DATA8. How would we respond to breach?

7. What could go wrong?

Name: John Smith                   ID: 123456. What are the risks?

Age: 35                            Age: 355. How is it protected?

ZIP: 90210                         Region: West4. How long will we keep it?

Income: $85,000                    Income bracket: $75K-$100K3. Who will have access?

2. Why do we need it?

✗ Can be re-identified            ✓ Cannot be traced to individual1. What personal data will we collect?

✗ Private information at risk     ✓ Privacy protected

```QUESTIONS TO ASK

```

---

### Privacy Impact Assessment

## Privacy Risks in AI

## Privacy in Practice

### Risk 1: Data Breaches

---

```

ATTACK VECTOR| **LGPD** | Brazil | Data subject rights, privacy | 2% revenue up to $50M |

  ↓| **PIPEDA** | Canada | Consent, security, breach notification | $100K fine |

Unauthenticated API Access| **CCPA** | California | Consumer rights, opt-out | $7,500 per violation |

  ↓| **GDPR** | Europe | Right to know, access, delete | 4% revenue or €20M |

Steal training data|---|---|---|---|

  ↓| Law | Region | Key Requirements | Penalty |

Extract personal information

  ↓### Major Privacy Laws

Sell on dark web

```## Regulatory Compliance



**Real Example:** OpenAI memory feature exposed conversations between users---



### Risk 2: Model Inversion```

☐ Incident Plan: Procedures if breach occurs

Attackers can reverse-engineer training data from model outputs.☐ Access Logging: Track who accessed what

☐ Retention Limits: Delete data after use period

```☐ Data Inventory: Know what data you have where

Training Data         AI Model         Attacker☐ Consent Forms: Users understand and agree

(Private)      →     (Public)   →     Reconstructs☐ Privacy Policy: Clear terms for data use

                                      personal data```

```

### Organizational Controls

### Risk 3: Privacy Leakage

| **Federated Learning** | Train models without centralizing data | Strong |

Models can memorize and inadvertently reveal training data.| **Differential Privacy** | Add mathematical noise to protect individuals | Strong |

| **Anonymization** | Remove identifying information | Moderate |

```| **Access Control** | Only authorized users access | Strong |

Large Language Model trained on:| **Encryption** | Data unreadable without key | Strong |

  • Public documents|---|---|---|

  • Web scrapes| Control | How It Works | Strength |

  • User-contributed data

  ↓### Technical Controls

Model can be tricked into repeating private information

```## Privacy Protection Strategies


---

# 🔎 Bias in AI: How It Distorts Decision-Making

## What Is Bias in AI?

**Bias** in AI occurs when algorithms make systematically unfair decisions that disadvantage certain groups due to:
- Imbalanced training data
- Flawed algorithms
- Proxy variables (hidden discrimination)
- Societal biases embedded in data

### Types of Bias



| Bias Type | Source | Impact | Example |

|---|---|---|---|

| **Data Bias** | Training data underrepresents groups | Model ignores minority patterns | Fewer women in training data → worse predictions for women |✨ **Acknowledge bias, address it systematically, and commit to fairness**

| **Algorithmic Bias** | Algorithm structure inherently unfair | Built-in discrimination | Algorithm weights certain features unfairly |

| **Proxy Bias** | Using correlated variables for discrimination | Indirect discrimination | ZIP code proxies for race |- **Choice to invest in fairness** → More equitable systems

| **Feedback Bias** | Previous biased decisions become training data | Bias compounds over time | Biased arrests → biased predictions → more biased arrests |- **Choice to ignore disparities** → Bias perpetuates

- **Choice to use biased data** → Bias in outputs

---

**Bias in AI is not inevitable—it's a choice:**

## How Bias Affects Decision-Making

## Key Takeaway

### 1. **Produces Systematically Unfair Outcomes**

---

```

BIASED OUTCOMES```

✓ Willingness to retire biased systems

Group A Approval: 90%✓ Transparency about limitations

Group B Approval: 60%✓ Stakeholder engagement

✓ Diverse teams building systems

Disparity: 30 percentage points✓ Governance and accountability

Legal threshold: <20% disparity```

→ Potential discrimination lawsuit

```### Layer 4: Organizational Level



### 2. **Creates Compounding Disadvantage**```

✓ Create appeals process

```✓ Include diverse reviewers

CYCLE OF BIAS✓ Maintain human oversight

✓ Alert if disparity increases

Past Discrimination✓ Monitor fairness metrics continuously

     ↓```

Biased Historical Data

     ↓### Layer 3: Deployment Level

AI Learns Discrimination

     ↓```

Biased AI Decisions✓ Use adversarial debiasing

     ↓✓ Adjust decision thresholds

Outcomes Become New Data✓ Test for discrimination

     ↓✓ Use fair algorithms

Next Model Inherits Bias✓ Apply fairness constraints

     ↓```

Disadvantage Compounds

```### Layer 2: Algorithm Level



### 3. **Reinforces Societal Inequalities**```

✓ Audit for historical biases

```✓ Document data sources and quality

SOCIETAL HARM✓ Remove proxy variables

✓ Balance underrepresented groups

Historical inequality + Biased AI✓ Ensure diverse representation

     ↓```

Wide-scale automated discrimination

     ↓### Layer 1: Data Level

Affected group worse off than before

     ↓## Addressing Bias: Multi-Layered Strategy

Inequality INCREASES

---

AI was supposed to be objective—instead it amplified bias

``````
