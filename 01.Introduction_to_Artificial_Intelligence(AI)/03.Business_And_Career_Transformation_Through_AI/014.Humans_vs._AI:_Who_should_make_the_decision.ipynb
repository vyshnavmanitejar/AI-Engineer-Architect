This lecture explores the question of who should make decisions—humans, artificial intelligence (AI), or a combination of both—and how to optimize decision-making effectiveness.

Decision-Making Performance Curves

AI systems tend to have high accuracy when confident in their predictions, especially at very high or very low confidence scores.
Humans perform better than AI when the AI is uncertain (around 50% confidence), as humans can use additional context and reasoning.
Augmented Intelligence: Combining Human and AI

Augmented intelligence leverages both AI and human judgment, often outperforming either alone in many cases.
The effectiveness of augmented intelligence depends on how AI recommendations are presented to humans to avoid cognitive biases.
Human Cognitive Bias and Presentation Effects

Forced display of AI recommendations can lead to automation bias, where humans overly rely on AI and ignore their own judgment.
Optional display, where humans can request AI input, helps reduce automation bias by allowing independent human evaluation first.
Trust issues arise when AI shows accuracy percentages, as humans may distrust recommendations that admit uncertainty.
Overall, the best decision-making approach depends on the confidence level of AI predictions and how human-AI collaboration is structured to minimize bias and maximize accuracy.




What is automation bias in human-AI decision making?

Automation bias in human-AI decision making is the tendency for humans to overly trust and favor suggestions from automated systems, such as AI recommendations, while ignoring or undervaluing their own judgment or contradictory information. This often happens when AI recommendations are presented to humans by default (forced display), leading people to assume the AI is always correct and potentially making poorer decisions as a result.




What is the role of confidence scores in AI decision making?

Confidence scores in AI decision making represent how certain the AI is about a particular prediction or decision. They play a key role by:

Indicating the AI's level of certainty, from 0% (definitely not) to 100% (definitely yes).
Helping determine when AI decisions are likely to be accurate (high confidence) or uncertain (mid-range confidence).
Guiding whether a decision should be handled by AI alone, a human, or a combination (augmented intelligence).
Influencing how humans interpret and trust AI recommendations, though showing accuracy percentages can sometimes reduce human trust.
Confidence scores help balance decision-making between AI and humans for better outcomes.




How can augmented intelligence improve fraud detection accuracy?

Augmented intelligence improves fraud detection accuracy by combining the strengths of both AI and human analysts:

AI handles alerts with high or low confidence scores, where it is statistically more accurate, reducing false positives and workload.
Human analysts focus on alerts with mid-range confidence scores, where AI is less certain, using their judgment, context, and additional information.
This collaboration leverages AI’s consistency and speed with human flexibility and reasoning.
Proper presentation of AI recommendations to humans minimizes cognitive biases, ensuring better decision quality.
Together, this approach leads to higher overall success rates in identifying real fraud cases.




How does confidence score affect AI and human decision making?

Confidence scores affect AI and human decision making in these ways:

High or Low Confidence Scores: AI predictions are usually very accurate when confidence is near 0% or 100%, so AI decisions are trusted and preferred here.
Mid-Range Confidence Scores: When AI confidence is around 50%, its accuracy drops, and humans tend to perform better by applying additional context and judgment.
Decision Allocation: Confidence scores help decide whether AI should make the decision, humans should intervene, or both should collaborate (augmented intelligence).
Human Trust: Confidence scores also influence how humans perceive and trust AI recommendations, sometimes reducing trust if AI openly shows uncertainty.
In summary, confidence scores guide who should decide and how decisions are made for better outcomes.
