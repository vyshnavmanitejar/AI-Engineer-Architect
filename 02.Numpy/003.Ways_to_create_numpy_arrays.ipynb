Numpy-3 : Ways to Create Numpy Arrays

```python
import numpy as np
```

From built-in python containers (Lists, Tuples etc)

```python
# using list
np. array([1,2,3,4])
```

```python
# using tuples
np.array( (1,2,3,4))
```

Upcasting in Numpy
- Upcasting in NumPy refers to the automatic conversion of data types when performing operations involving arrays of different types, so that no data is lost and the resuable has a compatible type that can hold all input values.

```python
np.array([1,2.5,3,4])
```

Two Dimensional numpy array

```python
np.array([[1,2,3],[4,5,6]])
```

Minimum dimensions 2

```python
np.array([[[1,2,3,4]]], ndmin = 2)
```

dtype

```python
np.array([1,'1',3,4], dtype = complex)
```

data consisting of more than one Data-type

```python
x = np. array([(1,2), (3,4.2) ],dtype=[( 'a', 'i4'), ('b',np. float32)])
x
```
