# 🔍 Python - Access List Items

## Introduction

Accessing list items is one of the most **fundamental operations** when working with Python lists. This notebook covers the various methods for retrieving and examining items using **indexing techniques**, **slicing**, and **membership testing**.

### Key Indexing Methods

- **Positive Indexing**: Access items from the beginning (`[0]`, `[1]`, etc.)
- **Negative Indexing**: Access items from the end (`[-1]`, `[-2]`, etc.)
- **Slicing**: Retrieve multiple items by specifying a range
- **Membership Testing**: Check if an item exists in the list

### Why This Matters

✅ **Extract specific data** from collections  
✅ **Modify or validate** particular items  
✅ **Iterate and search** efficiently  
✅ **Build logic** based on item presence  
✅ **Process data** from lists dynamically

## ✨ Accessing List Items Using Positive Indexes

**List indexing** allows you to access individual items by referring to their **position number** in square brackets. Understanding positive indexing is essential for working with lists.

### How Indexing Works

| Position | Item Value | Positive Index |
|----------|-----------|----------------|
| 1st | "apple" | `[0]` |
| 2nd | "banana" | `[1]` |
| 3rd | "cherry" | `[2]` |

### Key Concepts

- Items are indexed starting from **0** for the first item
- Each item occupies a unique index position in the list
- The index number is placed in square brackets `[]` after the list name
- Index must be within the valid range (0 to length-1)
- Indexes beyond the list length cause an **IndexError**

````python
# Example 1: Basic positive indexing
fruits = ["apple", "banana", "cherry"]
print("List:", fruits)
print("Item at index 0:", fruits[0])
print("Item at index 1:", fruits[1])
print("Item at index 2:", fruits[2])

# Example 2: Across different data types
mixed_list = [10, "hello", 3.14, True]
print("\nMixed list:", mixed_list)
print("First item (index 0):", mixed_list[0])
print("Fourth item (index 3):", mixed_list[3])

# Example 3: Accessing within a larger list
numbers = [100, 200, 300, 400, 500]
print("\nNumbers list:", numbers)
print("First:", numbers[0])
print("Middle:", numbers[2])
print("Last (using positive index):", numbers[4])
````

## 📋 Negative Indexing

**Negative indexing** allows you to access items starting from the **end** of the list, without needing to know the list length. This is particularly useful when you need recent or last items.

### How Negative Indexing Works

| Position | Item Value | Negative Index |
|----------|-----------|----------------|
| Last | "cherry" | `[-1]` |
| 2nd Last | "banana" | `[-2]` |
| 3rd Last | "apple" | `[-3]` |

### Key Points

- **-1** refers to the last item
- **-2** refers to the second last item
- **-3** refers to the third last item, and so on
- No need to calculate list length manually
- Useful for getting the most recent entry or working backwards

### Use Cases

✅ Get the **last element** without knowing the list size  
✅ Access **recent items** in logs or records  
✅ Process list items in **reverse order**  
✅ Implement **stack-like behavior**

````python
# Example 1: Basic negative indexing
fruits = ["apple", "banana", "cherry"]
print("List:", fruits)
print("Last item (index -1):", fruits[-1])
print("Second last item (index -2):", fruits[-2])
print("Third last item (index -3):", fruits[-3])

# Example 2: Negative indexing with larger list
fruits_large = ["apple", "banana", "cherry", "orange", "kiwi", "mango"]
print("\nLarger list:", fruits_large)
print("Last item:", fruits_large[-1])
print("2nd from last:", fruits_large[-2])
print("3rd from last:", fruits_large[-3])

# Example 3: Comparing positive and negative for same item
print("\nSame item, different indexes:")
print("fruits_large[0]:", fruits_large[0])  # Positive: first
print("fruits_large[-6]:", fruits_large[-6])  # Negative: first (6 items, so -6 = index 0)
````

## ✂️ Range of Indexes (List Slicing)

**List slicing** allows you to retrieve a **subset of items** by specifying a range. This creates a new list without modifying the original.

### Slicing Syntax

```python
list[start:end]
```

### Critical Rules

| Rule | Meaning | Example |
|------|---------|---------|
| `list[2:5]` | Start at 2, **exclude** 5 | Items at indices 2, 3, 4 |
| `list[:4]` | From beginning to 4 (excluded) | Items at indices 0, 1, 2, 3 |
| `list[2:]` | From index 2 to end | Items from 2 onwards |
| `list[:]` | Entire list (copy) | All items |

### Key Concepts

✅ **Start index is included**  
✅ **End index is excluded**  
✅ Returns a **new list** (doesn't modify original)  
✅ Safe slicing (out-of-range indices don't cause errors)

````python
# Example 1: Slice a range - from index 2 to 5 (5 excluded)
fruits = ["apple", "banana", "cherry", "orange", "kiwi", "melon", "mango"]
print("Original list:", fruits)
print("\nfruits[2:5] - items at indices 2, 3, 4:")
print(fruits[2:5])

# Example 2: Understanding the range
print("\nBreaking it down:")
print("  Index 2:", fruits[2])
print("  Index 3:", fruits[3])
print("  Index 4:", fruits[4])
print("  (Index 5 is NOT included)")

# Example 3: Try different ranges
print("\nOther ranges:")
print("fruits[0:3]:", fruits[0:3])
print("fruits[1:4]:", fruits[1:4])
print("fruits[3:6]:", fruits[3:6])
````

````python
# Example: Slice from beginning when start is omitted
fruits = ["apple", "banana", "cherry", "orange", "kiwi", "melon", "mango"]
print("Original list:", fruits)

# When start is omitted, it defaults to index 0
print("\nfruits[:4] - from beginning (index 0) to index 4 (excluded):")
print(fruits[:4])

# Equivalent to fruits[0:4]
print("\nfruits[0:4] - same result (explicit):")
print(fruits[0:4])

# More examples
print("\nOther examples with omitted start:")
print("fruits[:3]:", fruits[:3])
print("fruits[:5]:", fruits[:5])
print("fruits[:1]:", fruits[:1])  # Only first item
````

````python
# Example: Slice to end when end is omitted
fruits = ["apple", "banana", "cherry", "orange", "kiwi", "melon", "mango"]
print("Original list:", fruits)

# When end is omitted, it goes to the end of the list
print("\nfruits[2:] - from index 2 to the end:")
print(fruits[2:])

# More examples
print("\nOther examples with omitted end:")
print("fruits[0:]:", fruits[0:])   # Entire list
print("fruits[3:]:", fruits[3:])   # From index 3 onwards
print("fruits[5:]:", fruits[5:])   # Last 2 items
print("fruits[6:]:", fruits[6:])   # Just the last item
print("fruits[7:]:", fruits[7:])   # Empty (past the end)
````

## 🔄 Slicing with Negative Indexes

You can combine **negative indexes in slicing** to count from the end of the list, making it easy to extract items near the end without calculating lengths.

### Negative Slicing Examples

| Slice | Meaning | Result |
|-------|---------|--------|
| `list[-4:-1]` | From 4th-last to last (excluded) | Items 4, 3, 2 from end |
| `list[-3:]` | From 3rd-last to end | Last 3 items |
| `list[:-2]` | From beginning to 2nd-last (excluded) | All except last 2 |
| `list[-5:-2]` | Range near the end | Middle items (5th-last through 3rd-last) |

### Key Rules

- Same slicing syntax applies: `list[start:end]`
- Both indices count backwards from the end
- The start index is **included**, and the end index is **excluded**
- Useful for extracting **trailing items** without knowing list length

````python
# Example 1: Negative index slicing
fruits = ["apple", "banana", "cherry", "orange", "kiwi", "melon", "mango"]
print("Original list:", fruits)

# Indices:           0          1         2        3        4       5       6
# Negative:         -7         -6        -5       -4       -3      -2      -1

print("\nfruits[-4:-1] - from 4th-last to last (excluded):")
print(fruits[-4:-1])  # ["orange", "kiwi", "melon"]

# Example 2: Other negative slicing combinations
print("\nOther negative slicing examples:")
print("fruits[-3:]:", fruits[-3:])     # Last 3 items
print("fruits[:-2]:", fruits[:-2])     # All except last 2
print("fruits[-5:-2]:", fruits[-5:-2]) # Middle range from end

# Example 3: Using negative with positive (mixing)
print("\nMixing negative and positive indexes:")
print("fruits[1:-1]:", fruits[1:-1])   # From index 1 to 2nd-last (excluded)
````

## 🔎 Checking if an Item Exists in a List

The **`in`** keyword provides a simple way to test **membership**—whether a specific item exists in a list. This is essential for **validation** and **conditional logic**.

### Membership Testing

```python
item in list  # Returns True or False
```

### Key Features

✅ **Simple syntax**: `item in list`  
✅ **Returns Boolean**: `True` or `False`  
✅ **Case-sensitive**: `"Apple" in list` differs from `"apple" in list`  
✅ **Works with any data type**: strings, numbers, booleans, etc.  
✅ **Often combined with `not`**: `item not in list`

### Use Cases

✅ **Validate user input** before processing  
✅ **Check permissions** or allowed values  
✅ **Guard against duplicates** before adding  
✅ **Build conditional logic** based on presence  
✅ **Error prevention** in data processing

````python
# Example 1: Using 'in' to check item existence
fruits = ["apple", "banana", "cherry"]
print("List:", fruits)

if "apple" in fruits:
    print("✓ 'apple' is in the list")
else:
    print("✗ 'apple' is not in the list")

# Example 2: Checking for items that don't exist
print("\nChecking for 'grape':")
if "grape" in fruits:
    print("✓ 'grape' is in the list")
else:
    print("✗ 'grape' is not in the list")

# Example 3: Using 'not in'
print("\nUsing 'not in':")
if "mango" not in fruits:
    print("✓ Confirmed: 'mango' is NOT in the list")
````

````python
# Example: More practical membership testing scenarios

# Scenario 1: Validating allowed values
allowed_colors = ["red", "green", "blue", "yellow"]
user_input = "green"

print(f"Allowed colors: {allowed_colors}")
print(f"User chose: '{user_input}'")
if user_input in allowed_colors:
    print("✓ Valid choice!")
else:
    print("✗ Invalid choice! Please select from allowed colors.")

# Scenario 2: Checking multiple items
print("\n--- Checking multiple items ---")
shopping_list = ["milk", "eggs", "bread", "butter"]
print("Shopping list:", shopping_list)

items_to_buy = ["milk", "cheese", "eggs"]
for item in items_to_buy:
    if item in shopping_list:
        print(f"  ✓ {item} - already on list")
    else:
        print(f"  ✗ {item} - need to add to list")
````
