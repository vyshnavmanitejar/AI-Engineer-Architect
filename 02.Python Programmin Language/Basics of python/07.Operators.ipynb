# ⚙️ Python Operators

## Introduction

**Operators** are symbols in Python used to perform operations on variables and values. They allow you to manipulate data, compare values, make assignments, and control program flow.

### What is an Operator?

An operator is a special symbol that tells Python to perform a specific action on one or two values (called operands).

```python
10 + 5  # The + operator adds two numbers
x = 5   # The = operator assigns a value to a variable
```

### Key Concepts

✅ **Operands** - The values or variables being operated on  
✅ **Operator** - The symbol performing the action  
✅ **Expression** - A combination of operands and operators  

### Example

```python
print(10 + 5)  # Expression: 10 (operand) + (operator) 5 (operand) = 15
```

### Why Operators Matter

✅ Perform **mathematical calculations**  
✅ **Assign and modify** values  
✅ **Compare** values to create logic  
✅ **Combine conditions** for decision-making  
✅ Build **complex expressions** for programming

### Seven Main Operator Categories

| Category | Purpose | Example |
|----------|---------|---------|
| **Arithmetic** | Mathematical operations | `+`, `-`, `*`, `/` |
| **Assignment** | Assign values to variables | `=`, `+=`, `-=` |
| **Comparison** | Compare two values | `==`, `!=`, `>`, `<` |
| **Logical** | Combine Boolean conditions | `and`, `or`, `not` |
| **Identity** | Compare object identity | `is`, `is not` |
| **Membership** | Check membership in sequences | `in`, `not in` |
| **Bitwise** | Perform bit-level operations | `&`, `\|`, `^`, `~` |

````python
# Example 1: Basic arithmetic operator
print("Arithmetic operator (+):")
print(10 + 5)  # Output: 15

# Example 2: Operator with variables and values
print("\nOperations with variables:")
sum1 = 100 + 50      # 150 (100 + 50)
sum2 = sum1 + 250    # 400 (150 + 250)
sum3 = sum2 + sum2   # 800 (400 + 400)
print(f"sum1 = {sum1}")
print(f"sum2 = {sum2}")
print(f"sum3 = {sum3}")

# Example 3: Different types of operators
print("\nDifferent operators:")
x = 10 + 5      # Addition operator
print(f"10 + 5 = {x}")
y = 10 - 3      # Subtraction operator
print(f"10 - 3 = {y}")
z = 10 > 5      # Comparison operator (returns Boolean)
print(f"10 > 5 = {z}")
````

## 🔢 Arithmetic Operators

**Arithmetic operators** perform mathematical operations on numeric values.

### Available Arithmetic Operators

| Operator | Name | Purpose | Example | Result |
|----------|------|---------|---------|--------|
| `+` | Addition | Add two values | `10 + 5` | `15` |
| `-` | Subtraction | Subtract one value from another | `10 - 3` | `7` |
| `*` | Multiplication | Multiply two values | `10 * 5` | `50` |
| `/` | Division | Divide one value by another | `10 / 3` | `3.333...` |
| `//` | Floor Division | Divide and round down | `10 // 3` | `3` |
| `%` | Modulus | Return remainder of division | `10 % 3` | `1` |
| `**` | Exponentiation | Raise to power | `2 ** 3` | `8` |

### Key Points

✅ **Division (`/`)** always returns a **float**  
✅ **Floor division (`//`)** returns an **integer** (rounds down)  
✅ **Modulus (`%`)** returns the **remainder**  
✅ **Exponentiation (`**`)** for powers  

### Use Cases

✅ Mathematical calculations  
✅ Data processing and transformations  
✅ Creating formulas  
✅ Checking even/odd (modulus)

````python
# Example 1: Basic arithmetic operators
x = 10
y = 3
print("Arithmetic Operators:")
print(f"{x} + {y} = {x + y}")       # Addition
print(f"{x} - {y} = {x - y}")       # Subtraction
print(f"{x} * {y} = {x * y}")       # Multiplication
print(f"{x} / {y} = {x / y}")       # Division (returns float)

# Example 2: Floor division, modulus, exponentiation
print(f"\n{x} // {y} = {x // y}")   # Floor division
print(f"{x} % {y} = {x % y}")       # Modulus (remainder)
print(f"{x} ** {y} = {x ** y}")     # Exponentiation

# Example 3: Division returns float
print(f"\nType of 10 / 3: {type(10 / 3)}")    # <class 'float'>
print(f"Type of 10 // 3: {type(10 // 3)}")   # <class 'int'>

# Example 4: Using modulus to check even/odd
numbers = [10, 11, 12, 13, 14]
print("\nEven numbers (remainder 0):")
for num in numbers:
    if num % 2 == 0:
        print(f"  {num} is even")

# Example 5: Mathematical expressions
radius = 5
pi = 3.14159
area = pi * radius ** 2
print(f"\nCircle area (r={radius}): {area:.2f}")
````

## 📝 Assignment Operators

**Assignment operators** assign values to variables. They can also combine assignment with other operations.

### Available Assignment Operators

| Operator | Purpose | Example | Equivalent |
|----------|---------|---------|-----------|
| `=` | Assign | `x = 5` | - |
| `+=` | Add and assign | `x += 3` | `x = x + 3` |
| `-=` | Subtract and assign | `x -= 3` | `x = x - 3` |
| `*=` | Multiply and assign | `x *= 3` | `x = x * 3` |
| `/=` | Divide and assign | `x /= 3` | `x = x / 3` |
| `//=` | Floor divide and assign | `x //= 3` | `x = x // 3` |
| `%=` | Modulus and assign | `x %= 3` | `x = x % 3` |
| `**=` | Exponentiate and assign | `x **= 3` | `x = x ** 3` |

### Key Differences

✅ **`=`** - Simple assignment  
✅ **`+=`, `-=`, etc.** - Compound assignment (shorten and combine operations)  
✅ **Compound operators** are more efficient and readable  

### Why Use Compound Operators?

✅ **Shorter code** - `x += 1` vs `x = x + 1`  
✅ **More readable** - intent is clearer  
✅ **Works with any type** - strings, lists, numbers  
✅ **Common in loops** - frequent pattern in programming

````python
# Example 1: Simple assignment
x = 5
print(f"Simple assignment: x = {x}")

# Example 2: Arithmetic compound assignments
x = 10
print(f"\nStarting value: x = {x}")
x += 3
print(f"After x += 3: x = {x}")  # Same as x = x + 3
x -= 2
print(f"After x -= 2: x = {x}")  # Same as x = x - 2
x *= 2
print(f"After x *= 2: x = {x}")  # Same as x = x * 2

# Example 3: Division-based assignments
y = 20
print(f"\nStarting value: y = {y}")
y /= 4
print(f"After y /= 4: y = {y}")   # Division (returns float)
y = 20
y //= 4
print(f"After y //= 4: y = {y}")  # Floor division (returns int)

# Example 4: Modulus and exponentiation assignments
z = 10
z %= 3
print(f"\nAfter z %= 3: z = {z}")  # z = 1

z = 2
z **= 5
print(f"After z **= 5: z = {z}")  # z = 32

# Example 5: Assignment with other types
text = "Hello"
print(f"\nString assignment: text = '{text}'")
text += " World"  # String concatenation
print(f"After text += ' World': '{text}'")

items = [1, 2, 3]
print(f"\nList assignment: items = {items}")
items += [4, 5]  # List concatenation
print(f"After items += [4, 5]: {items}")
````

## 🔀 Comparison Operators

**Comparison operators** compare two values and return a **Boolean result** (`True` or `False`).

### Available Comparison Operators

| Operator | Name | Purpose | Example | Result |
|----------|------|---------|---------|--------|
| `==` | Equal to | Check if values are equal | `10 == 10` | `True` |
| `!=` | Not equal to | Check if values are different | `10 != 5` | `True` |
| `>` | Greater than | Check if left is greater | `10 > 5` | `True` |
| `<` | Less than | Check if left is smaller | `10 < 5` | `False` |
| `>=` | Greater or equal | Check if greater or equal | `10 >= 10` | `True` |
| `<=` | Less or equal | Check if less or equal | `10 <= 5` | `False` |

### Important Notes

⚠️ **Use `==` for comparison**, not `=` (assignment)  
✅ Comparison operators return **Boolean values**  
✅ Work with **any comparable types** (numbers, strings, etc.)  
✅ Foundation for **conditional logic** and decision-making

### Use Cases

✅ Build conditional statements (`if`, `elif`, `else`)  
✅ Create loop conditions  
✅ Validate and compare data  
✅ Build complex logic with multiple conditions

````python
# Example 1: Equality operators
x = 10
y = 10
z = 5
print("Equality operators:")
print(f"x == y: {x == y}")  # True (equal)
print(f"x != z: {x != z}")  # True (not equal)

# Example 2: Greater/Less than
print("\nComparison operators:")
print(f"x > z: {x > z}")    # True
print(f"x < z: {x < z}")    # False

# Example 3: Greater/Less than or equal
print(f"\nx >= y: {x >= y}")  # True (equal)
print(f"x <= z: {x <= z}")   # False

# Example 4: Comparing strings
name1 = "Alice"
name2 = "Bob"
print(f"\nString comparison:")
print(f"'{name1}' == '{name2}': {name1 == name2}")
print(f"'{name1}' != '{name2}': {name1 != name2}")

# Example 5: Using comparison in if statements
age = 18
print(f"\nUsing in if statement:")
if age >= 18:
    print(f"Age {age}: You are an adult")
else:
    print(f"Age {age}: You are a minor")

# Example 6: Comparing different types (be careful!)
print(f"\nType differences:")
print(f"10 == '10': {10 == '10'}")  # False (different types)
print(f"10 == 10.0: {10 == 10.0}")  # True (both numeric)
````

## 🔗 Logical Operators

**Logical operators** combine multiple Boolean expressions to create more complex conditions. (Covered in detail in the Booleans notebook)

### Available Logical Operators

| Operator | Purpose | Returns True When | Example |
|----------|---------|-------------------|---------|
| `and` | Logical AND | **Both** conditions are True | `a and b` |
| `or` | Logical OR | **At least one** condition is True | `a or b` |
| `not` | Logical NOT | The condition is False (negation) | `not a` |

### Quick Truth Tables

**AND:** Both must be True
- `True and True` → `True`
- `True and False` → `False`

**OR:** At least one must be True
- `True or False` → `True`
- `False or False` → `False`

**NOT:** Negate the result
- `not True` → `False`
- `not False` → `True`

### Use Cases

✅ Combine multiple conditions  
✅ Verify all requirements met (AND)  
✅ Check if any requirement met (OR)  
✅ Negate conditions (NOT)  
✅ Complex validation logic

````python
# Example 1: AND operator (both must be True)
x = 10
y = 5
print("AND operator:")
print(f"x > 5 and y > 0: {x > 5 and y > 0}")  # True (both true)
print(f"x > 5 and y > 10: {x > 5 and y > 10}")  # False (second false)

# Example 2: OR operator (at least one must be True)
print("\nOR operator:")
print(f"x > 15 or y > 0: {x > 15 or y > 0}")   # True (second true)
print(f"x < 5 or y < 0: {x < 5 or y < 0}")     # False (both false)

# Example 3: NOT operator (negation)
condition = True
print(f"\nNOT operator:")
print(f"not {condition}: {not condition}")    # False
print(f"not False: {not False}")              # True

# Example 4: Complex conditions with AND/OR
age = 25
has_license = True
has_insurance = False
print(f"\nComplex conditions:")
can_drive = age >= 18 and has_license
print(f"can_drive (age >= 18 AND license): {can_drive}")

can_get_loan = age >= 18 or has_insurance
print(f"can_get_loan (age >= 18 OR insurance): {can_get_loan}")

# Example 5: Practical use in if statements
score = 85
grade_level = 10
print(f"\nPractical example:")
if score >= 80 and grade_level >= 9:
    print("Eligible for honors program")
elif score >= 70 or grade_level >= 11:
    print("Eligible for advanced classes")
else:
    print("Standard class placement")
````

## 🆔 Identity Operators

**Identity operators** compare whether two variables refer to the **same object in memory** (not just if they have equal values).

### Available Identity Operators

| Operator | Purpose | Returns True When | Example |
|----------|---------|-------------------|---------|
| `is` | Identity check | Objects are **same object** in memory | `x is y` |
| `is not` | Negated identity | Objects are **different objects** | `x is not y` |

### Key Difference: `==` vs `is`

- **`==`** - Checks if values are equal
- **`is`** - Checks if they're the exact same object in memory

```python
x = [1, 2, 3]
y = [1, 2, 3]
print(x == y)   # True (equal values)
print(x is y)   # False (different objects)
```

### Common Use Cases

✅ Check if variable is **`None`**  
✅ Check if two variables point to **same object**  
✅ Memory and performance optimization  
✅ Comparing singleton objects  

### Note

Most common: `if x is None:` instead of `if x == None:`

````python
# Example 1: is vs == with lists
x = [1, 2, 3]
y = [1, 2, 3]
z = x

print("Identity vs Equality:")
print(f"x == y: {x == y}")   # True (same values)
print(f"x is y: {x is y}")   # False (different objects)
print(f"x is z: {x is z}")   # True (same object)

# Example 2: Checking for None (most common use)
value = None
print(f"\nChecking for None:")
print(f"value is None: {value is None}")      # True
print(f"value is not None: {value is not None}")  # False

value = "hello"
print(f"value is None: {value is None}")      # False
print(f"value is not None: {value is not None}")  # True

# Example 3: Identity with numbers (Python caches small integers)
a = 256
b = 256
c = 257
d = 257

print(f"\nIdentity with numbers:")
print(f"256 is 256: {a is b}")  # True (Python caches -5 to 256)
print(f"257 is 257: {c is d}")  # False (not cached)

# Example 4: Using is with custom checks
def check_validity(data):
    if data is None:
        return False
    if data is not "":
        return True
    return False

print(f"\nUsing in functions:")
print(f"None is valid: {check_validity(None)}")
print(f"'hello' is valid: {check_validity('hello')}")
````

## 🎯 Membership Operators

**Membership operators** check whether a value is found in a **sequence** (like lists, strings, tuples, sets, dictionaries).

### Available Membership Operators

| Operator | Purpose | Returns True When | Example |
|----------|---------|-------------------|---------|
| `in` | Membership check | Value is **in** the sequence | `"a" in "apple"` |
| `not in` | Negated membership | Value is **not in** the sequence | `"z" in "apple"` |

### Works With All Sequences

✅ **Strings** - `"a" in "apple"`  
✅ **Lists** - `1 in [1, 2, 3]`  
✅ **Tuples** - `"x" in ("x", "y", "z")`  
✅ **Sets** - `5 in {1, 2, 3, 4, 5}`  
✅ **Dictionaries** - `"key" in {"key": "value"}`  

### Use Cases

✅ Check if item exists in list  
✅ Validate character in string  
✅ Check if key exists in dictionary  
✅ Membership validation  
✅ Filtering and searching

````python
# Example 1: in operator with strings
text = "apple"
print("Membership in strings:")
print(f"'a' in '{text}': {'a' in text}")      # True
print(f"'z' in '{text}': {'z' in text}")      # False

# Example 2: in operator with lists
fruits = ["apple", "banana", "cherry"]
print(f"\nMembership in lists:")
print(f"'apple' in {fruits}: {'apple' in fruits}")  # True
print(f"'orange' in {fruits}: {'orange' in fruits}")  # False

# Example 3: in operator with other sequences
numbers = [1, 2, 3, 4, 5]
print(f"\nMembership in different types:")
print(f"3 in {numbers}: {3 in numbers}")      # True (list)
print(f"3 in (1, 2, 3): {3 in (1, 2, 3)}")    # True (tuple)
print(f"3 in {{1, 2, 3}}: {3 in {1, 2, 3}}")  # True (set)

# Example 4: Dictionary membership (checks keys)
person = {"name": "Alice", "age": 25, "city": "NYC"}
print(f"\nMembership in dictionaries:")
print(f"'name' in person: {'name' in person}")      # True (key exists)
print(f"'Alice' in person: {'Alice' in person}")    # False (value, not key)
print(f"'age' in person: {'age' in person}")        # True (key exists)

# Example 5: not in operator
print(f"\nnot in operator:")
print(f"'z' not in '{text}': {'z' not in text}")    # True
print(f"'orange' not in {fruits}: {'orange' not in fruits}")  # True

# Example 6: Using in loops and conditions
print(f"\nPractical use:")
password = "12345"
if "123" in password:
    print("Password contains sequential numbers - weak password")

for fruit in fruits:
    if len(fruit) > 5:
        print(f"  {fruit} has more than 5 letters")
````

## 🔧 Bitwise Operators (Advanced)

**Bitwise operators** perform operations on binary representations of integers. They work at the bit level and are **less commonly used** in foundation-level programming.

### Available Bitwise Operators

| Operator | Name | Purpose | Example | Result |
|----------|------|---------|---------|--------|
| `&` | AND | Bitwise AND | `5 & 3` | `1` |
| `\|` | OR | Bitwise OR | `5 \| 3` | `7` |
| `^` | XOR | Bitwise XOR (exclusive OR) | `5 ^ 3` | `6` |
| `~` | NOT | Bitwise NOT (invert) | `~5` | `-6` |
| `<<` | Left shift | Shift bits left | `5 << 1` | `10` |
| `>>` | Right shift | Shift bits right | `5 >> 1` | `2` |

### When to Use

✅ **Rarely used** for foundation-level programming  
✅ Advanced algorithms and optimization  
✅ Working with flags and permissions  
✅ Low-level programming and system work  
✅ Performance-critical code  

### Note

Bitwise operators are **advanced** and typically not needed when learning Python basics. They're included here for completeness.

````python
# Example 1: Bitwise AND, OR, XOR
a = 5     # Binary: 0101
b = 3     # Binary: 0011
print("Bitwise operators:")
print(f"{a} & {b} = {a & b}")   # AND: 0001 = 1
print(f"{a} | {b} = {a | b}")   # OR:  0111 = 7
print(f"{a} ^ {b} = {a ^ b}")   # XOR: 0110 = 6

# Example 2: Bitwise NOT
print(f"\nBitwise NOT:")
print(f"~{a} = {~a}")   # Inverts all bits

# Example 3: Bit shift operations
print(f"\nBit shift operations:")
print(f"{a} << 1 = {a << 1}")   # Left shift (multiply by 2)
print(f"{a} >> 1 = {a >> 1}")   # Right shift (divide by 2)

# Example 4: Practical use - checking if number is power of 2
def is_power_of_two(n):
    return n > 0 and (n & (n - 1)) == 0

print(f"\nChecking if power of 2:")
print(f"is_power_of_two(8): {is_power_of_two(8)}")   # True
print(f"is_power_of_two(7): {is_power_of_two(7)}")   # False

# Example 5: Using bitwise for flags
READ = 1      # 001
WRITE = 2     # 010
EXECUTE = 4   # 100

permissions = READ | WRITE  # User has read and write
print(f"\nUsing bitwise for permissions:")
print(f"Has READ: {bool(permissions & READ)}")      # True
print(f"Has WRITE: {bool(permissions & WRITE)}")    # True
print(f"Has EXECUTE: {bool(permissions & EXECUTE)}")  # False
````
