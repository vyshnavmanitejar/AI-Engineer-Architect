## 🐉 What is Python?

### Key Facts
- **High-level, interpreted** programming language
- **Dynamically typed** (flexible type system)
- Created by **Guido van Rossum** (late 1980s)
- **Easy to learn, powerful to use**

### Why Python?
✅ Simple and readable syntax  
✅ Extensive libraries and frameworks  
✅ Community-driven development  
✅ Cross-platform compatibility

## ⚙️ Interpreter vs Compiler

| Aspect | 📖 Interpreter | 🏗️ Compiler |
|--------|---------------|------------|
| **Function** | Executes code line by line | Converts entire code to machine language first |
| **Execution Speed** | Slower execution | Faster execution |
| **Development Cycle** | Faster development | Slower development |
| **Language Examples** | Python, JavaScript, Ruby | C, C++, Java, Go |
| **Error Detection** | Errors found during execution | Errors found during compilation |
| **Memory Usage** | Requires interpreter at runtime | Standalone executable |

## 📝 Syntax

**Syntax** = The grammar of programming  
A set of rules that defines how programs must be structured.

### Key Points
- **Incorrect syntax** → Code won't run (SyntaxError)
- **Like grammar rules** for sentences in a program
- **Every language** has different syntax rules
- **Governs structure** of every code statement

```
Example: Python requires colons after if statements
```

---

## 💡 Semantics

**Semantics** = The meaning of code

Even with **correct syntax**, semantics can be wrong!

### Example of Semantic Error
```python
# Syntax is correct, but semantically wrong:
result = 10 / 0  # ❌ Division by zero
```
