# 🛠️ Python String Methods

## Introduction

**String methods** are built-in functions attached to string objects that allow you to perform common operations on strings like searching, replacing, formatting, and case conversion.

### Key Characteristics

✅ **Built-in** - Available on all string objects  
✅ **Non-destructive** - Return new strings (strings are immutable)  
✅ **Chainable** - Can combine multiple methods  
✅ **Foundation-critical** - Essential for text processing  

### Important Note

**All string methods return new values** - they do **NOT** modify the original string (strings are immutable in Python).

```python
original = "hello"
result = original.upper()
# original is still "hello"
# result is "HELLO"
```

### Method Categories

String methods can be organized into four main groups:

| Category | Purpose | Examples |
|----------|---------|----------|
| **Case Modification** | Change letter case | `upper()`, `lower()`, `title()` |
| **Searching & Finding** | Locate substrings | `find()`, `count()`, `startswith()` |
| **String Manipulation** | Modify content | `replace()`, `split()`, `strip()` |
| **Formatting & Joining** | Combine and format | `join()`, `format()`, `center()` |

---

## 🔤 Case Modification Methods

These methods change the **case (uppercase/lowercase)** of letters in a string.

### Available Methods

| Method | Purpose | Returns | Example |
|--------|---------|---------|---------|
| `upper()` | Convert all to uppercase | String | `"hello".upper()` → `"HELLO"` |
| `lower()` | Convert all to lowercase | String | `"HELLO".lower()` → `"hello"` |
| `title()` | Capitalize first letter of each word | String | `"hello world".title()` → `"Hello World"` |
| `capitalize()` | Capitalize only first letter | String | `"hello world".capitalize()` → `"Hello world"` |
| `swapcase()` | Swap upper/lower cases | String | `"HeLLo".swapcase()` → `"hEllO"` |
| `casefold()` | Convert to lowercase (advanced) | String | `"ß".casefold()` → Similar to `lower()` |

### Use Cases

✅ Normalizing user input for comparison  
✅ Formatting titles and headers  
✅ Creating consistent output  
✅ Data validation and standardization  
✅ Preparing text for processing

````python
# Example 1: upper() - convert to uppercase
text = "hello world"
print(f"Original: {text}")
print(f"Upper: {text.upper()}")

# Example 2: lower() - convert to lowercase
text = "HELLO WORLD"
print(f"\nOriginal: {text}")
print(f"Lower: {text.lower()}")

# Example 3: title() - capitalize first letter of each word
text = "hello world from python"
print(f"\nOriginal: {text}")
print(f"Title: {text.title()}")

# Example 4: capitalize() - capitalize only first letter
text = "hello world from python"
print(f"\nOriginal: {text}")
print(f"Capitalize: {text.capitalize()}")

# Example 5: swapcase() - swap cases
text = "HeLLo WoRLd"
print(f"\nOriginal: {text}")
print(f"Swapcase: {text.swapcase()}")
````

## 🔎 Searching & Finding Methods

These methods **locate substrings** within a string and return information about their position or existence.

### Available Methods

| Method | Purpose | Returns | Example |
|--------|---------|---------|---------|
| `find()` | Search for substring, return index | Integer (-1 if not found) | `"hello".find("l")` → `2` |
| `index()` | Search for substring, return index | Integer (raises error if not found) | `"hello".index("l")` → `2` |
| `count()` | Count occurrences of substring | Integer | `"hello".count("l")` → `2` |
| `startswith()` | Check if string starts with substring | Boolean | `"hello".startswith("he")` → `True` |
| `endswith()` | Check if string ends with substring | Boolean | `"hello".endswith("lo")` → `True` |
| `rfind()` | Search from right, return last index | Integer (-1 if not found) | `"hello".rfind("l")` → `3` |
| `rindex()` | Search from right, return last index | Integer (raises error if not found) | `"hello".rindex("l")` → `3` |

### Use Cases

✅ Validating string content  
✅ Finding positions of substrings  
✅ Checking string patterns  
✅ Counting occurrences  
✅ Data validation and verification

````python
# Example 1: find() - search for substring
text = "hello world"
position = text.find("world")
print(f"Text: {text}")
print(f"Position of 'world': {position}")
print(f"Not found result: {text.find('xyz')}")

# Example 2: count() - count occurrences
text = "banana"
count = text.count("a")
print(f"\nText: {text}")
print(f"Count of 'a': {count}")

# Example 3: startswith() and endswith() - check beginning/end
text = "hello world"
print(f"\nText: {text}")
print(f"Starts with 'hello': {text.startswith('hello')}")
print(f"Ends with 'world': {text.endswith('world')}")

# Example 4: rfind() - find last occurrence
text = "hello hello hello"
last_position = text.rfind("hello")
print(f"\nText: {text}")
print(f"First 'hello' at: {text.find('hello')}")
print(f"Last 'hello' at: {last_position}")

# Example 5: Practical use - validating email
email = "user@example.com"
if email.startswith("user") and "@" in email and email.endswith(".com"):
    print(f"\nEmail '{email}' looks valid")
````

## ✂️ String Manipulation Methods

These methods **modify, split, or clean** strings to change their structure and content.

### Available Methods

| Method | Purpose | Returns | Example |
|--------|---------|---------|---------|
| `replace()` | Replace substring with another | String | `"hello".replace("l", "x")` → `"hexxo"` |
| `split()` | Split string by separator | List | `"a,b,c".split(",")` → `["a", "b", "c"]` |
| `rsplit()` | Split from right | List | `"a,b,c".rsplit(",", 1)` → `["a,b", "c"]` |
| `splitlines()` | Split at line breaks | List | `"a\nb\nc".splitlines()` → `["a", "b", "c"]` |
| `join()` | Join list elements into string | String | `",".join(["a", "b"])` → `"a,b"` |
| `strip()` | Remove whitespace both ends | String | `"  hello  ".strip()` → `"hello"` |
| `rstrip()` | Remove whitespace right end | String | `"hello  ".rstrip()` → `"hello"` |
| `lstrip()` | Remove whitespace left end | String | `"  hello".lstrip()` → `"hello"` |
| `center()` | Center string with padding | String | `"hello".center(11)` → `"   hello   "` |
| `partition()` | Split into three parts | Tuple | `"hello".partition("l")` → `("he", "l", "lo")` |

### Use Cases

✅ Data cleaning and normalization  
✅ Parsing structured text  
✅ Creating aligned output  
✅ Combining multiple strings  
✅ Text formatting

````python
# Example 1: replace() - replace substrings
text = "hello world world"
result = text.replace("world", "python")
print(f"Original: {text}")
print(f"Replaced: {result}")

# Example 2: split() - split into list
text = "apple,banana,cherry"
items = text.split(",")
print(f"\nOriginal: {text}")
print(f"Split result: {items}")

# Example 3: join() - join list into string
items = ["apple", "banana", "cherry"]
result = ", ".join(items)
print(f"\nItems: {items}")
print(f"Joined: {result}")

# Example 4: strip() - remove whitespace
text = "  hello world  "
print(f"\nOriginal: '{text}'")
print(f"After strip(): '{text.strip()}'")
print(f"After rstrip(): '{text.rstrip()}'")

# Example 5: center() and partition()
text = "hi"
print(f"\nOriginal: '{text}'")
print(f"Centered: '{text.center(10)}'")
print(f"Partition 'l' in 'hello': {'hello'.partition('l')}")
````

## ⚙️ Checking & Boolean Methods

These methods **check conditions** on strings and return `True` or `False` (yes/no questions).

### Available Methods

| Method | Purpose | Returns | Example |
|--------|---------|---------|---------|
| `isalpha()` | Contains only letters | Boolean | `"abc".isalpha()` → `True` |
| `isdigit()` | Contains only digits | Boolean | `"123".isdigit()` → `True` |
| `isalnum()` | Contains only letters/digits | Boolean | `"abc123".isalnum()` → `True` |
| `isspace()` | Contains only whitespace | Boolean | `"   ".isspace()` → `True` |
| `islower()` | All letters are lowercase | Boolean | `"hello".islower()` → `True` |
| `isupper()` | All letters are uppercase | Boolean | `"HELLO".isupper()` → `True` |
| `istitle()` | First char of each word uppercase | Boolean | `"Hello World".istitle()` → `True` |
| `startswith()` | Starts with substring | Boolean | `"hello".startswith("he")` → `True` |
| `endswith()` | Ends with substring | Boolean | `"hello".endswith("lo")` → `True` |

### Use Cases

✅ Input validation  
✅ Type checking  
✅ Conditional logic  
✅ Data verification  
✅ String classification

````python
# Example 1: isalpha() and isdigit() - check content type
text1 = "hello"
text2 = "12345"
text3 = "hello123"

print(f"'{text1}' is alpha: {text1.isalpha()}")
print(f"'{text2}' is digit: {text2.isdigit()}")
print(f"'{text3}' is alnum: {text3.isalnum()}")

# Example 2: islower() and isupper() - check case
print(f"\n'{text1}' is lower: {text1.islower()}")
print(f"'{text2.upper()}' is upper: {text2.upper().isupper()}")

# Example 3: isspace() - all whitespace
whitespace = "   \t\n"
print(f"\nWhitespace string is space: {whitespace.isspace()}")

# Example 4: istitle() - check title case
titles = ["Hello World", "hello world", "HELLO WORLD"]
for title in titles:
    print(f"'{title}' is title: {title.istitle()}")

# Example 5: Practical use - input validation
password = "Pass123"
is_valid = len(password) >= 6 and password[0].isupper() and any(c.isdigit() for c in password)
print(f"\nPassword '{password}' valid: {is_valid}")
````

## 🔗 Chaining String Methods

Since string methods **return new strings**, you can **chain multiple methods** together for powerful text processing.

### Syntax

```
string.method1().method2().method3()
```

Each method works on the result of the previous method.

### Examples

**Clean and normalize user input:**
```python
user_input = "  HELLO WORLD  "
cleaned = user_input.strip().lower()
# Result: "hello world"
```

**Process CSV data:**
```python
data = "apple, BANANA, cherry"
items = [item.strip().lower() for item in data.split(",")]
# Result: ["apple", "banana", "cherry"]
```

**Format structured text:**
```python
text = "hello world"
result = text.upper().center(20).replace("O", "0")
# Result: formatted and replaced text
```

### Benefits of Chaining

✅ **Concise** - fewer intermediate variables  
✅ **Readable** - clear transformation pipeline  
✅ **Efficient** - each step is clear and logical  
✅ **Powerful** - combine operations easily

````python
# Example 1: Chain for cleaning user input
user_input = "  HELLO WORLD  "
cleaned = user_input.strip().lower()
print(f"Original: '{user_input}'")
print(f"Cleaned: '{cleaned}'")

# Example 2: Chain for text normalization
email = "  USER@EXAMPLE.COM  "
normalized = email.strip().lower()
print(f"\nOriginal: '{email}'")
print(f"Normalized: '{normalized}'")

# Example 3: Chain multiple transformations
text = "hello world"
result = text.upper().replace("O", "0").center(20)
print(f"\nOriginal: '{text}'")
print(f"Transformed: '{result}'")

# Example 4: Chain with split and join
csv_data = "  apple , BANANA , cherry  "
items = [item.strip().lower() for item in csv_data.split(",")]
cleaned = ", ".join(items)
print(f"\nOriginal: '{csv_data}'")
print(f"Cleaned: '{cleaned}'")

# Example 5: Real-world example - process user input for filename
user_name = "  John's New File  "
filename = user_name.strip().lower().replace(" ", "_").replace("'", "")
print(f"\nUser input: '{user_name}'")
print(f"Filename: '{filename}.txt'")
````

### Quick Reference - Most Common Methods

| Task | Method | Example |
|------|--------|---------|
| Convert case | `upper()`, `lower()`, `title()` | `"hello".upper()` |
| Clean whitespace | `strip()`, `rstrip()`, `lstrip()` | `"  hi  ".strip()` |
| Find substring | `find()`, `count()` | `"hello".find("l")` |
| Replace text | `replace()` | `"old".replace("o", "0")` |
| Split text | `split()`, `splitlines()` | `"a,b".split(",")` |
| Join items | `join()` | `",".join(["a", "b"])` |
| Check content | `isalpha()`, `isdigit()` | `"123".isdigit()` |

### Method Categories by Purpose

**For Data Cleaning:**
- `strip()`, `lower()`, `replace()` - normalize input
- Chain them: `text.strip().lower().replace(...)`

**For Text Processing:**
- `split()`, `join()` - parse and reconstruct
- `replace()` - modify content
- `find()`, `count()` - locate substrings

**For Validation:**
- `isalpha()`, `isdigit()`, `isalnum()` - type checking
- `startswith()`, `endswith()` - pattern matching
- `isupper()`, `islower()` - case checking

**For Formatting:**
- `upper()`, `lower()`, `title()` - case conversion
- `center()` - add padding
- `partition()` - split carefully

### Critical Rules

🔹 **Remember: Strings are immutable** - methods return new strings  
🔹 **Must assign result**: `result = text.upper()` (not `text.upper()`)  
🔹 **Can chain methods**: `text.strip().lower().replace(...)`  
🔹 **Methods are case-sensitive**: `"Hello".find("h")` returns -1  
🔹 **Use appropriate method**: `find()` for searching, `startswith()` for checking  

### Common Method Combinations
