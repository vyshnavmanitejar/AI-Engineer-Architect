## 🎯 What is String Slicing?

**String slicing** is a technique to **extract a portion** (substring) of a string using **indices**.

### Characteristics
- Extract **part of a string** without modifying the original
- Use **colon syntax** `[start:end]`
- Works with **positive and negative indices**
- **End index is excluded** (not included)
- Very useful for string manipulation

### Why String Slicing?
✅ Extract specific parts of text  
✅ Split strings into components  
✅ Process user input  
✅ Work with data without modification  

---

## 📍 Basic Slicing Syntax

### The Slice Notation
```python
string[start:end]
```

- **`start`**: Index where slice begins (included)
- **`end`**: Index where slice ends (NOT included)
- **`start` is optional** (defaults to 0)
- **`end` is optional** (defaults to end of string)

### Important Rules

| Rule | Example | Result |
|------|---------|--------|
| **End is NOT included** | `"hello"[1:4]` | `"ell"` (not including index 4) |
| **Zero-based indexing** | `"hello"[0:2]` | `"he"` |
| **Works both ways** | `"hello"[2:4]` | `"ll"` |

### Visual Example
```
String:  "H  e  l  l  o"
Indices:  0  1  2  3  4
          
"Hello"[1:4] → "ell" (positions 1, 2, 3 - NOT 4)
"Hello"[0:2] → "He" (positions 0, 1)
```

````python
# Basic String Slicing Examples

# Example 1: Basic slicing from position 2 to 5
text = "Hello, World!"
print("Example 1 - Basic Slicing:")
print(f"String: '{text}'")
print(f"text[2:5] = '{text[2:5]}'")
print(f"text[0:5] = '{text[0:5]}'")
print(f"text[7:12] = '{text[7:12]}'")
print()

# Example 2: Understanding the slice
word = "Python"
print("Example 2 - Understanding Slices:")
print(f"String: '{word}'")
for i in range(len(word)):
    print(f"Index {i}: '{word[i]}'")
print()
print(f"word[1:4] = '{word[1:4]}' (positions 1, 2, 3 - NOT 4)")
print(f"word[0:2] = '{word[0:2]}'")
print(f"word[3:6] = '{word[3:6]}'")
print()

# Example 3: Different slices from same string
message = "Programming"
print("Example 3 - Different Slices:")
print(f"String: '{message}'")
print(f"message[0:7] = '{message[0:7]}'")
print(f"message[3:9] = '{message[3:9]}'")
print(f"message[4:12] = '{message[4:12]}'")
````

## ⬅️ Slice From the Start

By **omitting the start index**, Python starts from **index 0** (beginning of string).

### Syntax
```python
string[:end]
```

- **No start index** specified = start from position 0
- **End index** determines where to stop
- Remember: **end index is NOT included**

### Examples

| Code | Result | Description |
|------|--------|-------------|
| `"Hello"[:2]` | `"He"` | First 2 characters |
| `"Hello"[:3]` | `"Hel"` | First 3 characters |
| `"Hello"[:5]` | `"Hello"` | All characters |
| `"Hello"[:]` | `"Hello"` | Entire string |

### Use Cases
✅ Get first N characters  
✅ Extract prefix or header  
✅ Get substring from beginning  

### Important
- **`:5` is the same as `[0:5]`**
- **`[:]` returns a copy of entire string**
- **Extremely common in text processing**

````python
# Slicing From the Start Examples

# Example 1: Omitting start index
text = "Hello, World!"
print("Example 1 - Slicing from the start (omitting start index):")
print(f"String: '{text}'")
print(f"text[:5] = '{text[:5]}'")
print(f"text[:7] = '{text[:7]}'")
print(f"text[:13] = '{text[:13]}'")
print()

# Example 2: Getting first N characters
word = "Programming"
print("Example 2 - Getting first N characters:")
for n in [1, 3, 5, 8, 11]:
    print(f"First {n} characters: '{word[:n]}'")
print()

# Example 3: Practical use cases
email = "john.doe@example.com"
domain_start = email.find("@")
print("Example 3 - Practical example (extract username from email):")
print(f"Email: {email}")
print(f"Username: {email[:domain_start]}")
print()

# Example 4: Getting full string
data = "Important Data"
copy = data[:]
print("Example 4 - Get full string copy:")
print(f"Original: '{data}'")
print(f"Copy (using [:]): '{copy}'")
print(f"Are they equal? {data == copy}")
````

## ➡️ Slice To the End

By **omitting the end index**, Python continues to the **end of the string**.

### Syntax
```python
string[start:]
```

- **Start index** determines where to begin
- **No end index** specified = go to end
- **All characters from start onwards** are included

### Examples

| Code | Result | Description |
|------|--------|-------------|
| `"Hello"[2:]` | `"llo"` | From position 2 to end |
| `"Hello"[1:]` | `"ello"` | From position 1 to end |
| `"Hello"[0:]` | `"Hello"` | From position 0 to end |
| `"Hello"[5:]` | `""` | Empty (start is past end) |

### Use Cases
✅ Get last N characters (with length)  
✅ Extract suffix or tail content  
✅ Skip first N characters  
✅ Get substring from middle to end  

### Common Patterns
```python
text = "Hello, World!"
text[7:]    # Get "World!"
text[1:]    # Skip first character
text[:-1]   # All except last (using negative)
```

````python
# Slicing To the End Examples

# Example 1: Omitting end index
text = "Hello, World!"
print("Example 1 - Slicing to the end (omitting end index):")
print(f"String: '{text}'")
print(f"text[0:] = '{text[0:]}'")
print(f"text[2:] = '{text[2:]}'")
print(f"text[7:] = '{text[7:]}'")
print(f"text[8:] = '{text[8:]}'")
print()

# Example 2: Skipping first N characters
word = "Programming"
print("Example 2 - Skipping first N characters:")
for n in [0, 3, 6, 9]:
    print(f"Skip first {n}: '{word[n:]}'")
print()

# Example 3: Practical use - file extensions
filename = "document.pdf"
print("Example 3 - Practical example (extract file extension):")
dot_position = filename.find(".")
extension = filename[dot_position+1:]
print(f"Filename: {filename}")
print(f"Extension: {extension}")
print()

# Example 4: Extract domain from email
email = "user@example.com"
at_position = email.find("@")
print("Example 4 - Extract domain from email:")
print(f"Email: {email}")
print(f"Domain: {email[at_position+1:]}")
````

## ⬅️➡️ Negative Index Slicing

Use **negative indices** to slice from the **end of the string** backwards.

### How Negative Indexing Works

**Negative indices count from the end:**
- `-1` = last character
- `-2` = second-to-last character
- `-5` = fifth character from the end

### Visual Reference
```
String:  "P  r  o  g  r  a  m"
Positive: 0  1  2  3  4  5  6
Negative: -7 -6 -5 -4 -3 -2 -1
```

### Slicing with Negative Indices

| Code | Result | Description |
|------|--------|-------------|
| `"Program"[-3:]` | `"ram"` | Last 3 characters |
| `"Program"[:-2]` | `"Progra"` | All except last 2 |
| `"Program"[-5:-2]` | `"gra"` | From position -5 to -2 |
| `"Program"[-6:-1]` | `"rogra"` | Middle section |

### Use Cases
✅ Get last N characters  
✅ Remove last N characters  
✅ Work from the end without knowing length  
✅ Extract file extensions easily  

### Important
- **Negative slicing is very powerful** for text processing
- **`[:-1]` gets all except the last character**
- **`[-3:]` gets the last 3 characters**
- **No need to know string length**

````python
# Example 1: Get last 3 characters
text = "Python"
last_three = text[-3:]
print(f"Get last 3 characters from '{text}': {last_three}")

# Example 2: Remove last 2 characters
email = "user@example.com!"
clean_email = email[:-1]
print(f"Remove last character from '{email}': {clean_email}")

# Example 3: Negative range - middle section
word = "Programming"
middle = word[-7:-2]
print(f"Get characters from position -7 to -2 in '{word}': {middle}")

# Example 4: Extract file extension efficiently
filename = "document.pdf"
extension = filename[-3:]
print(f"File extension of '{filename}': {extension}")
````

## 🎯 Key Takeaways

You now understand:
✅ Basic slicing with `[start:end]`  
✅ Slicing from the start with `[:end]`  
✅ Slicing to the end with `[start:]`  
✅ Negative indexing and slicing `[-start:-end]`  

### Quick Reference Table

| Syntax | Example | Result | Purpose |
|--------|---------|--------|---------|
| `[start:end]` | `"Python"[0:3]` | `"Pyt"` | Get characters from start to end (end excluded) |
| `[:end]` | `"Python"[:4]` | `"Pyth"` | Get first N characters |
| `[start:]` | `"Python"[3:]` | `"hon"` | Get from position to end |
| `[-n:]` | `"Python"[-3:]` | `"hon"` | Get last N characters |
| `[:-n]` | `"Python"[:-2]` | `"Pyth"` | Get all except last N |

### Common Patterns

**Get first 5 characters**
```python
text[:5]
```

**Get last 3 characters**
```python
text[-3:]
```

**Remove last character**
```python
text[:-1]
```

**Skip first and last character**
```python
text[1:-1]
```

### Important Notes
- **Index starts at 0**, not 1
- **End index is excluded** (`[0:3]` gets indices 0, 1, 2 only)
- **Negative indices count from the end** (-1 is last, -2 is second-to-last)
- **Slicing doesn't modify original string** (strings are immutable)
- **Out-of-range indices don't cause errors** (safe slicing)
