NoSQL Databases

- Not-only SQL (NoSQL) databases are not relational, very more in structure.
- Depending on application, may perform more quickly or reduce technical overhead.
- Most NoSQL databeses store data in JSON format.
- Examples of NoSQL databeses:
    - Document databases: mongoDB, couchDB
    - Key-value stores: Riak, Voldemort, Redis
    - Graph databases: Neo4j, HyperGraph
    - Wide-cilumn stores: Cassandra, HBase
