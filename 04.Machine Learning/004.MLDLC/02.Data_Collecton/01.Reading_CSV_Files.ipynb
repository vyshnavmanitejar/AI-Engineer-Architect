```python
import pandas as pd
filepath = '/Users/vyshnav/AI Engineer:Architect/Machine Learning/004.MLDLC/02.Data_Collecton/Iris.csv'

# import the data
data = pd.read_csv(filepath)

# print few rows of the data
print(data.head())
```

```python
# Read csv files: useful arguments

# Different delimiters - tab-separated file (.tsv):
data = pd.read_csv(filepath, sep='\t')

# Different delimeters - space-separated file:
data = pd.read_csv(filepath, delim_whitespace=True)

# Don't use first row for column names:
data = pd.read_csv(filepath, header=None)

# Specify column names:
data = pd.read_csv(filepath, names=['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'species'])

# Custom missing value indicators:
data = pd.read_csv(filepath, na_values=['NA', 'N/A', 'na', 'n/a', '--'])
```
